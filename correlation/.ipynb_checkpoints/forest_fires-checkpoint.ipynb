```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression

font_size = 30
font_size_ticks = 25

data_all = pd.read_csv ("../data/forestfires.csv")
data = data_all[ (data_all["area"]>10) & (data_all["area"] < 100)]
print(data.columns)
print(len(data["area"]))
plt.figure()
plt.hist((data["area"]),bins=50)
```

```python
data.corr()
```

```python
area = data["area"]
humidity = data["RH"]
temperature = data["temp"]
```

```python
fig = plt.figure(figsize = (9,9)) 
plt.scatter(humidity,area,c="black",s=60,cmap="gray",edgecolors='black',rasterized=True)
#plt.xlabel(r"$a$", fontsize=font_size,labelpad=10)
#h = plt.ylabel(r"$b$", fontsize=font_size,labelpad=15)
#h.set_rotation(0)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
#plt.ylim([ymin,ymax])
#plt.xlim([tmin,tmax])
# plt.xlabel('Supplement intake',fontsize=font_size,labelpad = 10)
# plt.ylabel('Weight change',fontsize=font_size,labelpad = 15)
# plt.savefig('plots/chipmunks_samples_'+str(ind)+'_allblack.pdf',bbox_inches="tight")

fig = plt.figure(figsize = (12,9)) 
im =plt.scatter(humidity,area,c=temperature,s=120,cmap="gray",edgecolors='black',rasterized=True)
#plt.xlabel(r"$a$", fontsize=font_size,labelpad=10)
#h = plt.ylabel(r"$b$", fontsize=font_size,labelpad=15)
#h.set_rotation(0)
cbar = plt.colorbar(im,shrink=0.8)
cbar.ax.tick_params(labelsize=font_size_ticks,pad = 20)
#cbar.set_label(label='Food intake',fontsize=font_size,labelpad = 20)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
#plt.ylim([ymin,ymax])
#plt.xlim([tmin,tmax])
#plt.xlabel('Supplement intake',fontsize=font_size,labelpad = 10)
#plt.ylabel('Weight change',fontsize=font_size,labelpad = 15)
#plt.savefig('plots/chipmunks_samples_'+str(ind)+'.pdf',bbox_inches="tight")
```

```python
standardized_temperature = (temperature.values - np.mean(temperature.values))/np.std(temperature.values)
standardized_humidity = (humidity.values - np.mean(humidity.values))/np.std(humidity.values)
standardized_area = (area.values - np.mean(area.values))/np.std(area.values)
X = np.array([[standardized_temperature,standardized_humidity]])[0].T
X.shape
reg = LinearRegression(fit_intercept=False).fit(X, area)
print(reg.coef_)
```
