```python
import os.path
import math
os.environ["PATH"] += os.pathsep + '/usr/local/texlive/2019/bin/x86_64-darwin'
print(os.getenv("PATH"))
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['text.usetex'] = True
from numpy.random import default_rng
rng = default_rng(2023)

font_size = 45
font_size_ticks = 45

tmin = -3.99
tmax = 3.99
cmin = -3.99
cmax = 3.99
ymin = -3.99
ymax = 3.99
t = np.linspace(tmin, tmax, 200)
c = np.linspace(cmin, cmax, 200)
y = np.linspace(ymin, ymax, 200)

C,T = np.meshgrid(c,t)
Y,T = np.meshgrid(y,t)
```

```python
font_size_ticks_smaller = 40
E_po_tc = T

plt.figure(figsize = (9,9)) 
im = plt.imshow(E_po_tc, cmap='gray', extent=[tmin, tmax,cmin, cmax], origin='lower')
plt.xticks(fontsize=font_size_ticks_smaller) 
plt.yticks(fontsize=font_size_ticks_smaller)
ax = plt.gca()
ax.tick_params(axis='both', which='major', pad=15)
cbar = plt.colorbar(im,shrink=0.8)
cbar.ax.tick_params(labelsize=font_size_ticks,pad = 20)
cbar.set_label(label='Weight-change mean',fontsize=font_size,labelpad = 20)
plt.xlabel('Supplement intake',fontsize=font_size,labelpad = 10)
plt.ylabel('Food intake',fontsize=font_size,labelpad = 15)
plt.savefig('plots/guineapigs_po_tc.pdf',bbox_inches="tight")
plt.show()
```

```python
def gaussian(x,y,invSigma,Sigma):
        return (np.exp(-(invSigma[0,0] * x**2 + 2*invSigma[0,1]*x*y + invSigma[1,1] * y**2)/2)
                      /(2 * np.pi * np.sqrt(np.linalg.det(Sigma))))
    
def fmt(x):
    s = f"{x:.4f}"
    if s.endswith("0"):
        s = f"{x:.1f}"
    return rf"{s}" if plt.rcParams["text.usetex"] else f"{s}"

    
rho_tc_vals = [0.8,-0.8,0]

levels_weight_tc = [0.1,1,5,15,25]
levels_weight_yc = [0.1,1,5,10]

n=500
markersize = 10
```

```python
ind = 0
rho_tc = 0.8

Sigma_tc = np.array([[1,rho_tc],[rho_tc,1]])

invSigma_tc = np.linalg.inv(Sigma_tc)

z_tc = gaussian(T,C,invSigma_tc,Sigma_tc)

plt.figure(figsize=(9,9))
# plt.scatter(stature,weight,s=markersize,color="gray",alpha=alpha,rasterized=True)
CS = plt.contour(T,C, 1e2*z_tc,colors='black', levels=levels_weight_tc)
plt.clabel(CS, CS.levels, inline=True, fmt=fmt,fontsize=font_size_ticks)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([cmin,cmax])
plt.xlim([tmin,tmax])
plt.xlabel('Supplement intake',fontsize=font_size,labelpad = 10)
plt.ylabel('Food intake',fontsize=font_size,labelpad = 15)
plt.savefig('plots/guineapigs_tc_'+str(ind)+'.pdf',bbox_inches="tight")


Sigma_yc = np.array([[1,rho_tc],[rho_tc,2]])

invSigma_yc = np.linalg.inv(Sigma_yc)

z_yc = gaussian(T,C,invSigma_yc,Sigma_yc)

plt.figure(figsize=(9,9))
# plt.scatter(stature,weight,s=markersize,color="gray",alpha=alpha,rasterized=True)
CS = plt.contour(T,Y, 1e2*z_yc,colors='black', levels=levels_weight_yc)
plt.clabel(CS, CS.levels, inline=True, fmt=fmt,fontsize=font_size_ticks)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([ymin,ymax])
plt.xlim([tmin,tmax])
plt.xlabel('Supplement intake',fontsize=font_size,labelpad = 10)
plt.ylabel('Weight change',fontsize=font_size,labelpad = 15)
plt.savefig('plots/guineapigs_ty_'+str(ind)+'.pdf',bbox_inches="tight")

samples_tc = rng.multivariate_normal([0,0], Sigma_tc, size=n)
samples_t = samples_tc[:,0]
samples_c = samples_tc[:,1]
samples_y = rng.normal(loc=samples_c)
print("Correlation coefficient: " + str(np.corrcoef(samples_t,samples_y)))

fig = plt.figure(figsize = (9,9)) 
plt.scatter(samples_t,samples_y,c="black",s=60,cmap="gray",edgecolors='black',rasterized=True)
#plt.xlabel(r"$a$", fontsize=font_size,labelpad=10)
#h = plt.ylabel(r"$b$", fontsize=font_size,labelpad=15)
#h.set_rotation(0)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([ymin,ymax])
plt.xlim([tmin,tmax])
plt.xlabel('Supplement intake',fontsize=font_size,labelpad = 10)
plt.ylabel('Weight change',fontsize=font_size,labelpad = 15)
plt.savefig('plots/guineapigs_samples_'+str(ind)+'_allblack.pdf',bbox_inches="tight")

fig = plt.figure(figsize = (9,9)) 
im = plt.scatter(samples_t,samples_y,c=samples_c,s=120,cmap="gray",edgecolors='black',
                 vmin=-3,vmax=3,rasterized=True)
#plt.xlabel(r"$a$", fontsize=font_size,labelpad=10)
#h = plt.ylabel(r"$b$", fontsize=font_size,labelpad=15)
#h.set_rotation(0)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([ymin,ymax])
plt.xlim([tmin,tmax])
plt.xlabel('Supplement intake',fontsize=font_size,labelpad = 10)
plt.ylabel('Weight change',fontsize=font_size,labelpad = 15)
plt.savefig('plots/guineapigs_samples_'+str(ind)+'.pdf',bbox_inches="tight")

fig = plt.figure(figsize = (9,9)) 
cbar = plt.colorbar(im,shrink=1.1)
cbar.ax.tick_params(labelsize=font_size,pad = 10)
cbar.set_label(label='Food intake',fontsize=font_size+10,labelpad = 10)
ax = plt.gca()
ax.remove()
plt.savefig('plots/guineapigs_cbar_'+str(ind)+'.pdf',bbox_inches="tight")

fig = plt.figure(figsize = (9,9)) 
im = plt.scatter(samples_t,samples_c,c=samples_y,s=80,cmap="gray",edgecolors='black',
                 vmin=-3,vmax=3,rasterized=True)
#plt.xlabel(r"$a$", fontsize=font_size,labelpad=10)
#h = plt.ylabel(r"$b$", fontsize=font_size,labelpad=15)
#h.set_rotation(0)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([ymin,ymax])
plt.xlim([tmin,tmax])
plt.xlabel('Supplement intake',fontsize=font_size,labelpad = 10)
plt.ylabel('Food intake',fontsize=font_size,labelpad = 15)
plt.savefig('plots/guineapigs_samples_'+str(ind)+'_tc.pdf',bbox_inches="tight")

fig = plt.figure(figsize = (9,9)) 
cbar = plt.colorbar(im,shrink=1.1)
cbar.ax.tick_params(labelsize=font_size,pad = 10)
cbar.set_label(label='Weight change',fontsize=font_size+10,labelpad = 10)
ax = plt.gca()
ax.remove()
plt.savefig('plots/guineapigs_cbar_'+str(ind)+'_tc.pdf',bbox_inches="tight")

plt.figure(figsize=(9,9))
plt.scatter(samples_t,samples_y,s=40,color="gray",rasterized=True)
plt.plot(t,rho_tc * t,lw=3,
         color="black",ls="dashed",label="Linear model")
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([ymin,ymax])
plt.xlim([tmin,tmax])
plt.xlabel('Supplement intake',fontsize=font_size,labelpad = 10)
plt.ylabel('Weight change',fontsize=font_size,labelpad = 15)
plt.legend(fontsize=font_size,framealpha=1)
plt.savefig('plots/guineapigs_linear_estimator_yt_'+str(ind)+'.pdf',bbox_inches="tight")

fig = plt.figure(figsize = (15,12)) 
sc = plt.scatter(samples_t,samples_c,c=samples_y,s=80,cmap="gray",edgecolor='black',vmax = 3,vmin = -3,
            rasterized=True)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
ax = plt.gca()
ax.tick_params(axis='both', which='major', pad=15)
cbar = plt.colorbar(sc,shrink=1)
cbar.ax.tick_params(labelsize=font_size_ticks,pad = 20)
cbar.set_label(label='Weight change',fontsize=font_size,labelpad = 20)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([cmin,cmax])
plt.xlim([tmin,tmax])
plt.xlabel('Supplement intake',fontsize=font_size,labelpad = 20)
plt.ylabel('Food intake',fontsize=font_size,labelpad = 15)
plt.savefig('plots/guineapigs_linear_estimator_ytc_'+str(ind)+'_nocontours.pdf',bbox_inches="tight")
plt.show()

fig = plt.figure(figsize = (12,9)) 
sc = plt.scatter(samples_t,samples_c,c=samples_y,s=80,cmap="gray",edgecolor='black',vmax = 3,vmin = -3,
            rasterized=True)
CS = plt.contour(C.T,colors="black", extent=[tmin,tmax,cmin,cmax], 
                origin='lower',linewidths=3,linestyles="dashed",levels=5)
manual_locations = [(-1, 3),(-2,1.5),(-3,0),  (2, -1.5), (1, -3)]
plt.clabel(CS, fontsize=font_size, inline=1,fmt = '%1.1f',manual =manual_locations)#,ticks=Contourrange)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
ax = plt.gca()
ax.tick_params(axis='both', which='major', pad=15)
cbar = plt.colorbar(sc,shrink=1)
cbar.ax.tick_params(labelsize=font_size_ticks,pad = 20)
cbar.set_label(label='Weight change',fontsize=font_size,labelpad = 20)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([cmin,cmax])
plt.xlim([tmin,tmax])
plt.xlabel('Supplement intake',fontsize=font_size,labelpad = 20)
plt.ylabel('Food intake',fontsize=font_size,labelpad = 15)
plt.savefig('plots/guineapigs_linear_estimator_ytc_'+str(ind)+'.pdf',bbox_inches="tight")
plt.show()
```
