```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['text.usetex'] = True
from scipy.stats import norm

font_size = 30
font_size_legend = 22
font_size_ticks = 30

m = 4
pmf = np.zeros(m)
pmf[0] = 0.3
pmf[1] = 0.3
pmf[3] = 0.4

mean = np.sum(pmf * np.arange(0,m,1))
mean_square = np.sum(pmf * np.arange(0,m,1)**2)
variance = mean_square - mean**2
print(variance)
```

```python
n_conv = 9

pmf_conv_array = [pmf]
for ind in range(n_conv):
    pmf_conv = np.convolve(pmf_conv_array[ind],pmf,mode='full')
    pmf_conv_array.append(pmf_conv)
```

```python
xmin = -1
xmax = 28
ymax = 0.42
for ind in range(n_conv):
    pmf = pmf_conv_array[ind]
    inds_plot = np.arange(0,len(pmf))
    inds_plot_cont = np.linspace(xmin,xmax,500)
    
    plt.figure(figsize=(9,6))
    plt.plot(inds_plot[pmf>0],pmf[pmf>0],'o',markersize=10,markerfacecolor='white',markeredgecolor="black",markeredgewidth=2)
    #plt.xticks(np.arange(1, 6+1, 1))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks([0.1,0.2,0.3,0.4],fontsize=font_size_ticks)
    plt.ylim(0,ymax)
    plt.xlim(xmin,xmax)
    plt.ylabel(r"$p_{\tilde{s}}(s)$",fontsize=font_size,labelpad = 30)
    plt.xlabel(r"$s$ (total points)",fontsize=font_size)
    #plt.savefig('plots/die_histogram.pdf',bbox_inches="tight")
    
    n = ind + 1 
    mean_n = mean * n
    variance_n = variance * n
    
    plt.figure(figsize=(9,6))
    plt.plot(inds_plot[pmf>0],pmf[pmf>0],'o',markersize=10,markerfacecolor='white',markeredgecolor="gray",markeredgewidth=2)
    plt.plot(inds_plot_cont,norm.pdf(inds_plot_cont,mean_n,np.sqrt(variance_n)),lw=2.5,color="black",ls="dashed",
             label='Gaussian approximation')
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks([0.1,0.2,0.3,0.4],fontsize=font_size_ticks)
    plt.ylim(0,ymax)
    plt.ylabel(r"$p_{\tilde{s}}(s)$",fontsize=font_size,labelpad = 30)
    plt.xlabel(r"$s$ (total points)",fontsize=font_size)
    plt.legend(fontsize=font_size)
    #plt.savefig('plots/die_histogram.pdf',bbox_inches="tight")

```
