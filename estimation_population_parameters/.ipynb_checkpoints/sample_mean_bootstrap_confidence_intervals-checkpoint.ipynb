```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from scipy import stats
from scipy.stats import norm
from scipy.stats import gaussian_kde
from numpy.random import default_rng
import seaborn as sns
import matplotlib
matplotlib.rcParams['text.usetex'] = True
matplotlib.rc('font', family='serif', serif='cm10')

matplotlib.rc('text', usetex=True)
#matplotlib.rcParams['text.latex.preamble'] = r'\boldmath'
import matplotlib.transforms

font_size = 50
```

```python
font_size_legend = 50
font_size_ticks = 50

data = pd.read_csv ("../data/ANSUR II MALE Public.csv", encoding='latin-1')

stature = data['stature'] / 10.
N = len(stature)
population_mean = np.mean(stature)
population_std = np.std(stature)
print("N: " + str(len(stature)))
print("Population mean: " + str(population_mean))
print("Population standard deviation: " + str(population_std))
```

```python
rng = default_rng(1234)
tries_sm = int(1e6)
sample_means = np.zeros(tries_sm)
for ind_tries in range(tries_sm):
    samples = rng.choice(stature,n,replace=True)
    sample_means[ind_tries] = np.mean(samples)
```

```python
n_bins = 20
x_min = 169
x_max = 182.5
x = np.linspace(x_min,x_max,500)
y_max = 0.4
eps = 1e-3
x = np.linspace(x_min, x_max, 1000)

n = 25
    
fig = plt.figure(figsize=(15,9))
# plt.xticks([172,174,175.6,177,179],fontsize=font_size_ticks) 
# plt.plot(x,norm.pdf(x,population_mean,population_std/np.sqrt(n)),lw=2.5,color="black")
pdf = gaussian_kde(sample_means,bw_method=0.1)
plt.plot(x,pdf(x),lw=2.5,color="black")

# plt.fill_between(ind_interval,curve,color="lightgray",label="Probability = 0.95")
[l,u] = np.percentile(sample_means,[2.5, 97.5])
x_aux = x[x>l]
x_aux = x_aux[x_aux<u]
plt.fill_between(x=x_aux,y1=pdf(x_aux),color="lightgray",label="Probability = 0.95")
plt.ylim([0,y_max])
plt.xlim([x_min,x_max])
plt.yticks(np.arange(0,y_max,0.1),np.round(np.arange(0,y_max,0.1),1),fontsize=font_size_ticks)
plt.xticks([170,172.5,population_mean,180],[170,172.5,r"$\mu_{pop}$",180],fontsize=font_size)
ax = plt.gca()
plt.ylabel(r"Pdf of the sample mean",fontsize=font_size,labelpad=20)
#plt.xlabel('Height in cms',fontsize=font_size,labelpad=20)
plt.legend(fontsize=font_size_legend,loc='upper right')
#    plt.savefig('plots/height_gaussian_n_'+str(n) +'.pdf',bbox_inches="tight")
```

```python
rng = default_rng(123)
step = 0.4

n_bs = int(1e5)
tries = 3

plt.figure(figsize=(15,12))
#plt.xticks([170,172,174,175.6,178,180,182],[170,172,174,"\boldmath$\mu_{pop}",178,180,182],fontsize=font_size)
plt.xticks([170,172.5,population_mean,180],[170,172.5,r"$\mu_{pop}$",180],fontsize=font_size)
plt.xlabel('Height (cms)',fontsize=font_size,labelpad=20)
plt.yticks([]) 
#plt.yticks([170,172,174,175.6,178,180,182],fontsize=font_size_ticks)
plt.vlines(population_mean, -1,tries,lw=6,ls="dotted",color="black")
plt.xlim(x_min,x_max)
plt.ylim(-0.04,3 * step-0.04)
plt.ylabel(r"Pdf of bootstrap mean and bootstrap confidence intervals",fontsize=font_size,labelpad=90)
for ind in range(tries):
    samples = rng.choice(stature,n,replace=True)
    print(samples.shape)
    sample_mean = np.mean(samples)
    sample_mean_bs = np.zeros(n_bs)
    for ind_bs in range(n_bs):
        samples_bs = rng.choice(samples,n,replace=True)
        sample_mean_bs[ind_bs] = np.mean(samples_bs)
    pdf_bs = gaussian_kde(sample_mean_bs,bw_method=0.1)
    plt.plot(x,ind * step + pdf_bs(x),lw=2.5,color="black")

    [l,u] = np.percentile(sample_mean_bs,[2.5, 97.5])
    x_aux = x[x>l]
    x_aux = x_aux[x_aux<u]
    if ind == 0:
        plt.fill_between(x=x_aux,y1=ind * step,y2=ind * step + pdf_bs(x_aux),color="lightgray",label="Probability = 0.95")
    else:
        plt.fill_between(x=x_aux,y1=ind * step,y2=ind * step + pdf_bs(x_aux),color="lightgray")
    if u < population_mean or l > population_mean:
        plt.hlines(ind * step,l,u,lw=6,color="black",ls="dashed")
    else:
        plt.hlines(ind * step,l,u,lw=6,color="black")
    if ind == 0:
        plt.plot(sample_mean,ind * step,"o",markersize=25,markeredgewidth=3,markerfacecolor = "white",
                 markeredgecolor = "black",label="Sample mean")
    else:
        plt.plot(sample_mean,ind * step,"o",markersize=25,markeredgewidth=3,markerfacecolor = "white",
                 markeredgecolor = "black")
plt.legend(fontsize=font_size_legend-14,loc="upper right",framealpha=1)
# plt.savefig('plots/height_confidence_intervals_'+str(n) +'.pdf',bbox_inches="tight")
```
