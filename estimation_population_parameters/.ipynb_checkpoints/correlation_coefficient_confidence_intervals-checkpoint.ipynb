```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from scipy.stats import norm
from scipy import stats
import matplotlib
from numpy.random import default_rng
matplotlib.rcParams['text.usetex'] = True

alldata = pd.read_csv ("../data/ANSUR II MALE Public.csv", encoding='latin-1')

font_size = 35
font_size_legend = 35
font_size_ticks = 35

stature = alldata['stature'] / 10.
foot = alldata['footlength']/10.

stature_min = stature.min()
stature_max = stature.max()
foot_min = foot.min()
foot_max = foot.max()

data = np.stack((stature,foot))

rho_pop = np.corrcoef(data)[0,1]
print("Population correlation coefficient: " + str(rho_pop))
```

```python
rng = default_rng(2021)

n_vals = [25,100]

tries_plot = 3
markersize =4


for n in n_vals:

    for ind in range(tries_plot):
    
        indices = rng.choice(data.shape[1],n,replace=True)
        sample_rho = np.corrcoef(data[:,indices])[0,1]
        print(sample_rho)
        plt.figure(figsize=(9,9))
        plt.scatter(stature,foot,s=4,color="lightgray",rasterized=True)
        plt.scatter(stature[indices],foot[indices],s=16,color="black",rasterized=True,label='Samples')
        plt.xticks(fontsize=font_size_ticks) 
        plt.yticks(fontsize=font_size_ticks)
        plt.xticks(fontsize=font_size_ticks) 
        plt.yticks(fontsize=font_size_ticks)
        plt.ylim([foot_min,foot_max])
        plt.xlim([stature_min,stature_max])
        plt.xlabel('Height (cms)',fontsize=font_size,labelpad = 10)
        plt.ylabel('Foot length (cms)',fontsize=font_size,labelpad = 15)
        plt.legend(fontsize=font_size_legend,loc="lower right")
        plt.savefig('plots/rhos_sample_'+str(ind)+'_' + str(n) + '.pdf',bbox_inches="tight")
        plt.show()
```

```python
tries = int(1e6)

sample_rhos_list = []
for n in n_vals:
    print(n)
    sample_rhos = np.zeros(tries)
    for ind_tries in range(tries):
        indices = rng.choice(data.shape[1],n,replace=True)
        sample_rhos[ind_tries] = np.corrcoef(data[:,indices])[0,1]
    sample_rhos_list.append(sample_rhos)
```

```python
n_bins=400

x = np.linspace(0,1,1000)
x_min = 0.25
x_max = 1
y_max =  9.95#5.75

for ind_n,n in enumerate(n_vals):
    print(n)
    sample_rhos = sample_rhos_list[ind_n]
    mu = np.mean(sample_rhos)
    sigma = np.std(sample_rhos)

    plt.figure(figsize=(9,9))
    plt.xticks([0.1,0.3,0.5,0.718,1],[r'$0.1$',r'$0.3$',r'$0.5$',r'$\rho_{pop}$',r'$1.0$'],fontsize=font_size_ticks) 
    # plt.xticks(fontsize=font_size_ticks)
    plt.yticks(fontsize=font_size_ticks)
    plt.hist(sample_rhos,bins=n_bins,color='darkgray', edgecolor='darkgray',
                  linewidth=2,density=True)
    plt.plot(x,norm.pdf(x,mu,sigma),lw=2.5,color="black",label='Gaussian approximation')
    plt.xlim([x_min,x_max])
    plt.ylim([0,y_max])
    plt.ylabel(r"Probability density",fontsize=font_size,labelpad=20)
    plt.xlabel('Correlation coefficient',fontsize=font_size,labelpad=20)
    plt.legend(fontsize=font_size_legend,loc="upper left")
    plt.savefig('plots/rhos_population_'+ str(n) + '.pdf',bbox_inches="tight")
```

```python
def fisher_transformation(x):
    return 0.5 * np.log( (1+x) / (1-x))

def inverse_fisher_transformation(y):
    return (np.exp(2*y) - 1) / (np.exp(2*y) + 1)

sample_rhos = sample_rhos_list[0]

transformed_rhos = fisher_transformation(sample_rhos)
transformed_rho_pop = fisher_transformation(rho_pop)
print("Transformed population correlation coefficient: " + str(transformed_rho_pop))

x_min_fisher = 0 # 0.45
x_max_fisher = 1.75 # 1.35
x = np.linspace(x_min_fisher,x_max_fisher,1000)
y_max_fisher = 2.35

mu = np.mean(transformed_rhos)
sigma = np.std(transformed_rhos)
print("Mean: " + str(mu))
print("Standard error: " + str(sigma))
print("Approximation: " + str(1./np.sqrt(n-3)))

plt.figure(figsize=(9,9))
plt.xticks([0.25,0.5,0.9036,1.25,1.5],[r'$0.25$',r'$0.5$',r'$M(\rho_{pop})$',r'$1.25$',r'$1.5$'],fontsize=font_size_ticks) 
plt.xticks(fontsize=font_size_ticks)
plt.yticks(fontsize=font_size_ticks)
plt.hist(transformed_rhos,bins=n_bins,color='darkgray', edgecolor='darkgray',
              linewidth=2,density=True)
plt.plot(x,norm.pdf(x,mu,sigma),lw=2.5,color="black",label='Gaussian approximation')
plt.ylim([0,y_max_fisher])
plt.xlim([x_min_fisher,x_max_fisher])
plt.ylabel(r"Probability density",fontsize=font_size,labelpad=20)
plt.xlabel('Transformed correlation coefficient',fontsize=font_size,labelpad=20)
plt.legend(fontsize=font_size_legend,loc="upper left")
plt.savefig('plots/transformed_rhos_population.pdf',bbox_inches="tight")

y_max = 2.25
x_vals = np.linspace(0,1,1000)
transformed_vals = 0.5 * np.log( (1+x_vals) / (1-x_vals))
plt.figure(figsize=(9,9))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.plot(x_vals,transformed_vals,lw=3,color="black")
plt.xlim([0,1])
plt.ylim([-0.1,y_max])
plt.ylabel(r"$M(x)$",fontsize=font_size,labelpad=20)
plt.xlabel(r'$x$',fontsize=font_size,labelpad=20)
plt.savefig('plots/fisher_transformation.pdf',bbox_inches="tight")
```

```python
rng = default_rng(2023)

tries_bs = 5
markersize = 4
n_bs = int(1e5)

x_min = 0.05
x_max = 1
y_max = 12

n_bins_bs = 30

for ind_n,n in enumerate(n_vals):
    
    sample_rhos = sample_rhos_list[ind_n]
    
    for ind in range(tries_bs):

        indices = rng.choice(data.shape[1],n,replace=True)
        samples = data[:,indices]
        sample_rho = np.corrcoef(samples)[0,1]
        print("Sample correlation coefficient")
        print(sample_rho)
        plt.figure(figsize=(4.5,4.5))
        #plt.scatter(stature,foot,s=4,color="lightgray",rasterized=True)
        plt.scatter(stature[indices],foot[indices],s=16,color="black",rasterized=True,label='Samples')
        plt.xticks([150,170,190],fontsize=font_size_ticks) 
        plt.yticks(fontsize=font_size_ticks)
        plt.xticks(fontsize=font_size_ticks) 
        plt.yticks(fontsize=font_size_ticks)
        plt.ylim([foot_min,foot_max])
        plt.xlim([stature_min,stature_max])
        plt.xlabel('Height (cms)',fontsize=font_size,labelpad = 10)
        plt.ylabel('Foot length (cms)',fontsize=font_size,labelpad = 15)
        # plt.legend(fontsize=font_size_legend,loc="lower right")
        plt.savefig('plots/bootstrapping_rhos_sample_'+str(ind)+'_'+ str(n) + '.pdf',bbox_inches="tight")

        rhos_bs = np.zeros(n_bs)
        for ind_bs in range(n_bs):
            indices_bs = rng.choice(samples.shape[1],n,replace=True)
            samples_bs = samples[:,indices_bs]
            rhos_bs[ind_bs] = np.corrcoef(samples_bs)[0,1]
        print("Standard error estimate")
        print(np.std(rhos_bs))

        plt.figure(figsize=(9,9))
        #plt.xticks([0.6,0.718,0.8],[r'$0.6$',r'$\rho_{pop} = 0.718$',r'$0.8$'],fontsize=font_size_ticks)
        plt.xticks(fontsize=font_size_ticks)
        plt.yticks(fontsize=font_size_ticks)
        plt.hist(sample_rhos,bins=n_bins,color='gray', edgecolor='gray',
                      linewidth=2,density=True,label="True distribution")
        plt.hist(rhos_bs,bins=n_bins_bs,color='white', edgecolor='black',alpha=0.75,
                      linewidth=2,density=True,label="Bootstrap estimate")
        plt.xlim([x_min,x_max])
        plt.ylim([0,y_max])
        plt.ylabel(r"Probability density",fontsize=font_size,labelpad=20)
        plt.xlabel('Correlation coefficient',fontsize=font_size,labelpad=20)
        plt.legend(fontsize=font_size_legend,loc="upper left")
        plt.savefig('plots/bootstrapping_rhos_population_'+str(ind)+'_'+ str(n) + '.pdf',bbox_inches="tight")

    
```

```python
rng = default_rng(2023)
x_min = 0
x_max = 1

n_vals = [25,100] 
tries = 20
n_bs = int(1e5)

for ind_n,n in enumerate(n_vals):
    plt.figure(figsize=(15,15))
    plt.xticks([0,0.25,0.5,0.718,1],[r'$0$',r'$0.25$',r'$0.5$',r'$\rho_{pop}$',r'$1.0$'],fontsize=font_size) 
    plt.yticks([]) 
    #plt.yticks([170,172,174,175.6,178,180,182],fontsize=font_size_ticks)
    plt.vlines(rho_pop, -1,2*tries,lw=4,ls="dotted",color="black")
    plt.xlim(x_min,x_max)
    plt.ylim(-1, 2 * tries)
    plt.ylabel(r"Sample correlation coefficient",fontsize=font_size,labelpad=15)
    for ind in range(tries):
        print(ind)
        indices = rng.choice(data.shape[1],n,replace=True)
        samples = data[:,indices]
        sample_rho = np.corrcoef(samples)[0,1]
        rhos_bs = np.zeros(n_bs)
        for ind_bs in range(n_bs):
            indices_bs = rng.choice(samples.shape[1],n,replace=True)
            samples_bs = samples[:,indices_bs]
            rhos_bs[ind_bs] = np.corrcoef(samples_bs)[0,1]
        se_bs = np.std(rhos_bs)
        Phi_inv = 1.96 # Inverse cdf of 1-0.05/2
        l_norm = sample_rho - Phi_inv * se_bs
        u_norm = sample_rho + Phi_inv * se_bs
        
        if u_norm < rho_pop or l_norm > rho_pop:
            plt.hlines(2*ind+1,l_norm,u_norm,lw=6,color="gray",ls="dashed")
            print("Out!")
        else:
            if ind_n == 1 and ind == 0:
                plt.hlines(2*ind+1,l_norm,u_norm,lw=6,color="gray",label="Gaussian interval")
            else:
                plt.hlines(2*ind+1,l_norm,u_norm,lw=6,color="gray")
        plt.plot(sample_rho,2*ind+1,"o",markersize=15,markeredgewidth=3,markerfacecolor = "white",markeredgecolor = "black")
        
        [l_perc,u_perc] = np.percentile(rhos_bs,[2.5, 97.5])
        
        if u_perc < rho_pop or l_perc > rho_pop:
            plt.hlines(2*ind,l_perc,u_perc,lw=6,color="black",ls="dashed")
            print("Out!")
        else:
            if ind_n == 1 and ind == 0:
                plt.hlines(2*ind,l_perc,u_perc,lw=6,color="black",label="Percentile interval")
            else:
                plt.hlines(2*ind,l_perc,u_perc,lw=6,color="black")
        plt.plot(sample_rho,2*ind,"o",markersize=15,markeredgewidth=3,markerfacecolor = "white",markeredgecolor = "black")
    if ind_n == 1:
        plt.legend(fontsize=font_size_legend,loc="upper left")
    plt.savefig('plots/rhos_confidence_intervals_'+str(n) +'.pdf',bbox_inches="tight")
```
