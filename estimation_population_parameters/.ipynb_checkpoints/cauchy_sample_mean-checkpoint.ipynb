```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from scipy import stats
from scipy.stats import norm
from numpy.random import default_rng
rng = default_rng(2022)

font_size = 25
font_size_legend = 22
font_size_ticks = 22

fraction = 0.05
sigma = np.sqrt(fraction*(1-fraction))
```

```python
tries = 20#int(1e2)
n_vals = np.arange(10,1000,1)

ymin = 0
ymax = 0.15

plt.figure(figsize=(15,9))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim(ymin,ymax)
for ind_n,n in enumerate(n_vals):
    if n % 500 == 0:
        print(n)
    sample_means = (rng.binomial(n,fraction,tries)).astype('float')/n
    plt.plot(n * np.ones(tries),sample_means,'.',color="silver")
plt.semilogx(n_vals,fraction + sigma/np.sqrt(n_vals),lw=2,ls='dashed',color='black',
         label=r"$\theta_{pop}+\sqrt{\frac{ \theta_{pop} (1-\theta_{pop})}{n}}$")
plt.semilogx(n_vals,np.ones(len(n_vals))*fraction,lw=2,color='black',
         label=r"$\theta_{pop}$")
plt.semilogx(n_vals,fraction - sigma/np.sqrt(n_vals),lw=2,ls='dashed',color='black',
         label=r"$\theta_{pop}-\sqrt{\frac{ \theta_{pop} (1-\theta_{pop})}{n}}$")
plt.xlabel("n",fontsize=font_size)
plt.ylabel('Sample mean',fontsize=font_size,labelpad=20)
plt.legend(fontsize=font_size)
plt.savefig('plots/fraction_sample_mean_n.pdf',bbox_inches="tight")
```

```python
tries = int(1e4)
n_vals = [100,1000,10000]
n_bins = 20
x = np.arange(-4,4+1e-2,1e-2)
x_min = 0
x_max = 0.14
y_max = 190
x_min_norm = -4.1
x_max_norm = 4.1

for ind_n,n in enumerate(n_vals):
    sample_means = (rng.binomial(n,fraction,tries)).astype('float')/n  
    plt.figure(figsize=(9,9))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.hist(sample_means,bins=n_bins,color='white', edgecolor='black',
              linewidth=2,density=True)
    plt.xlim([x_min,x_max])
    #plt.ylim([0,y_max])
    plt.ylabel(r"Probability density of sample mean",fontsize=font_size,labelpad=20)
    #plt.xlabel('Height in cms',fontsize=font_size,labelpad=20)
    plt.savefig('plots/fraction_lln_'+str(n) +'.pdf',bbox_inches="tight")
    
    plt.figure(figsize=(9,9))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.hist((sample_means-fraction)/(sigma / np.sqrt(n)),bins=n_bins,color='white', edgecolor='black',
              linewidth=2,density=True)
    plt.plot(x,norm.pdf(x,0,1),lw=2.5,color="black",label='CLT approximation')
    plt.ylabel(r"Density of standardized sample mean",fontsize=font_size,labelpad=20)
    #plt.xlabel('Fraction',fontsize=font_size,labelpad=20)
    plt.legend(fontsize=font_size,loc='upper left')
    plt.ylim([0,0.6])
    plt.xlim([x_min_norm,x_max_norm])
    plt.savefig('plots/fraction_clt_'+str(n) +'.pdf',bbox_inches="tight")
```
