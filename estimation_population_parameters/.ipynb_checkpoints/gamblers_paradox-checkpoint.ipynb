```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from scipy import stats
from scipy.stats import norm
from numpy.random import default_rng
rng = default_rng(2021)

font_size = 25
font_size_legend = 22
font_size_ticks = 22
```

```python
tries = 20#int(1e2)
m = 200
n_vals = np.logspace(1,4,m,dtype=int)

ymin = -400
ymax = 400

plt.figure(figsize=(15,9))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim(ymin,ymax)
for ind_n,n in enumerate(n_vals):
    if n % 500 == 0:
        print(n)
    realizations = (np.random.standard_cauchy((n,tries))).astype('float')
    sample_means = np.sum(realizations,axis=0) / n
    plt.semilogx(n * np.ones(tries),sample_means,'.',color="black")
plt.xlabel("n",fontsize=font_size)
plt.ylabel('Sample mean',fontsize=font_size,labelpad=20)
#plt.legend(fontsize=font_size)
plt.savefig('plots/cauchy_sample_mean_n.pdf',bbox_inches="tight")
```
