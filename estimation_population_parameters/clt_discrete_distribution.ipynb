Synthetic example showing the convergence of the distribution of the sum of i.i.d. discrete random variables with finite variance to a Gaussian distribution\
Topics: Central limit theorem, convolution, sum of independent random variables

Author: Carlos Fernandez-Granda

```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['text.usetex'] = True
from scipy.stats import norm

font_size = 15
font_size_legend = 15
font_size_ticks = 15

m = 4
pmf = np.zeros(m)
pmf[0] = 0.3
pmf[1] = 0.3
pmf[3] = 0.4

mean = np.sum(pmf * np.arange(0,m,1))
mean_square = np.sum(pmf * np.arange(0,m,1)**2)
variance = mean_square - mean**2
```

```python
n_conv = 9

pmf_conv_array = [pmf]
for ind in range(n_conv):
    pmf_conv = np.convolve(pmf_conv_array[ind],pmf,mode='full')
    pmf_conv_array.append(pmf_conv)
```

```python
xmin = -1
xmax = 24.9
ymax = 0.42
for ind in range(n_conv):
    pmf = pmf_conv_array[ind]
    inds_plot = np.arange(0,len(pmf))
    inds_plot_cont = np.linspace(xmin,xmax,500)
    
    n = ind + 1
    mean_n = mean * n
    variance_n = variance * n
    
    plt.figure(figsize=(5,4))
    plt.plot(inds_plot[pmf>0],pmf[pmf>0],'o',markersize=6,markerfacecolor='black',markeredgecolor="black",markeredgewidth=2)
    plt.plot(inds_plot_cont,norm.pdf(inds_plot_cont,mean_n,np.sqrt(variance_n)),lw=2.5,color="black",ls="solid",
             label='Gaussian approximation')
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks([0.1,0.2,0.3,0.4],fontsize=font_size_ticks)
    plt.xlim(xmin,xmax)
    plt.ylim(0,ymax)
    if n == 1:
        plt.ylabel(r"$p_{\tilde{s}_1}(s)$",fontsize=font_size,labelpad = 5)
    elif n==2:
        plt.ylabel(r"$p_{\tilde{s}_2}(s)$",fontsize=font_size,labelpad = 5)
    elif n==3:
        plt.ylabel(r"$p_{\tilde{s}_3}(s)$",fontsize=font_size,labelpad = 5)
    elif n==4:
        plt.ylabel(r"$p_{\tilde{s}_4}(s)$",fontsize=font_size,labelpad = 5)
    elif n==5:
        plt.ylabel(r"$p_{\tilde{s}_5}(s)$",fontsize=font_size,labelpad = 5)
    elif n==6:
        plt.ylabel(r"$p_{\tilde{s}_6}(s)$",fontsize=font_size,labelpad = 5)
    elif n==7:
        plt.ylabel(r"$p_{\tilde{s}_7}(s)$",fontsize=font_size,labelpad = 5)
    elif n==8:
        plt.ylabel(r"$p_{\tilde{s}_8}(s)$",fontsize=font_size,labelpad = 5)
    elif n==9:
        plt.ylabel(r"$p_{\tilde{s}_9}(s)$",fontsize=font_size,labelpad = 5)
    plt.xlabel(r"$s$ (total points)",fontsize=font_size)
    plt.legend(fontsize=font_size_legend)
```
