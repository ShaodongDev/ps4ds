```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from scipy.stats import poisson
data = pd.read_csv ("./data/tsunamis.tsv", sep = '\t')
# data = pd.read_csv ("./data/earthquakes.tsv", sep = '\t')
# data = pd.read_csv ("./data/volcanoes.tsv", sep = '\t')
# print(data.columns)
years = data['Year']
min_year = min(years[1:])
max_year = 2020 # We don't consider 2021 because it is incomplete (this code was originally written in August 2021)
print('Earliest year: ' + str(min_year))
count = np.zeros(int(-min_year + 1 + max_year))
for year in years[1:]:
    if year != 2021:
        count[int(-min_year + year)] += 1
plt.figure(figsize=(16, 6))
plt.plot(np.arange(min_year,2021),count,'.')
plt.ylabel("Number of tsunamis")
plt.xlabel("Year")
plt.figure(figsize=(16, 6))
plt.plot(np.arange(min_year,2021)[-200:],count[-200:],'.')
plt.ylabel("Number of tsunamis")
plt.xlabel("Year")
```

```python
# We consider the period between 1850 and 2020, we will use 1850-1950 as a training test and 1951-2020 as a test set
training = count[-170:-70]
test = count[-70:-20]
plt.figure(figsize=(16, 6))
# sns.histplot(data=training, x="Number of tsunamis in year")
n, bins, patches = plt.hist(training, int(max(training)), density=False, facecolor='tomato', alpha=1,align='right')
```

```python
lambda_parameter = np.mean(training)
print("Estimated lambda parameter: " + str(lambda_parameter))
plt.figure(figsize=(16, 6))
n, bins, patches = plt.hist(training, int(max(training)), density=True, facecolor='tomato', alpha=1,align='right')
indices = range(int(max(training))+2)
plt.plot(indices, poisson.pmf(indices,lambda_parameter),'o',color='blue')                           
```

```python
plt.figure(figsize=(16, 6))
n, bins, patches = plt.hist(test, int(max(test)), density=True, facecolor='tomato', alpha=1,align='right')
n, bins, patches = plt.hist(training, int(max(training)), density=True, facecolor='cyan', alpha=0.5,align='right')
indices = range(int(max(training))+2)
plt.plot(indices, poisson.pmf(indices,lambda_parameter),'o',color='blue')
```
