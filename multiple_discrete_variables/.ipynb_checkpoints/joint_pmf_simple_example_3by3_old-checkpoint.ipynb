```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
matplotlib.rcParams['text.usetex'] = True

plt.close("all")
fsize = 30
fsize_3D = 20
fsize_label = 40

data = np.array([
[0.05,0.05,0.05],
[0.4,0.05,0.05],
[0.1,0.05,0.2],
])


column_names = [r'$1$',r'$2$',r'$3$']
row_names = column_names#  ['1','2','3']

fig = plt.figure(figsize=(12, 9))
ax = Axes3D(fig)

lx= len(data[0])            # Work out matrix dimensions
ly= len(data[:,0])
xpos = np.arange(0,lx,1)    # Set up a mesh of positions
ypos = np.arange(0,ly,1)
xpos, ypos = np.meshgrid(xpos, ypos)

xpos = xpos.flatten()   # Convert positions to 1D array
ypos = ypos.flatten()
zpos = np.zeros(lx*ly)

dx = 0.5 * np.ones_like(zpos)
dy = dx.copy()
dz = data.flatten()

cs = ['skyblue', 'skyblue', 'skyblue'] * ly
cs[2] = 'white'

ax.bar3d(xpos,ypos,zpos, dx, dy, dz, color=cs)

#sh()
ax.w_xaxis.set_ticks([0.25,1.25,2.25])
ax.w_yaxis.set_ticks([0.25,1.25,2.25])
ax.w_xaxis.set_ticklabels(column_names,fontsize=fsize)
ax.w_yaxis.set_ticklabels(row_names,fontsize=fsize)
ax.set_zticks([0, 0.1,0.2,0.3,0.4])
ax.w_zaxis.set_ticklabels([r'$0$', r'$0.1$',r'$0.2$',r'$0.3$',r'$0.4$'],fontsize=fsize_3D)
for t in ax.zaxis.get_major_ticks(): t.label.set_fontsize(fsize)
ax.set_xlabel(r'$a$',fontsize=fsize_label,labelpad=20)
ax.set_ylabel(r'$b$',fontsize=fsize_label,labelpad=20)
ax.zaxis.set_rotate_label(False)
ax.text(3.4, 0, 0.66, r'$p_{\tilde{a},\tilde{b}}(a,b)$', None,fontsize=fsize_label)
#ax.set_zlabel(r'$p_{\tilde{a},\tilde{b}}(a,b)$',fontsize=fsize_label,labelpad=20,
#              rotation=0)
#ax.set_title(r'$p_{\tilde{a},\tilde{b}}(a,b)$',fontsize=fsize_label,loc='right')
plt.xlim([0.05,2.5])
plt.ylim([0,2.45])
plt.show()
plt.tight_layout()
plt.savefig('plots/joint_pmf_8blue_1red.pdf',bbox_inches="tight")

cs = ['skyblue', 'skyblue', 'skyblue'] * ly

ax.bar3d(xpos,ypos,zpos, dx, dy, dz, color=cs)

#sh()
ax.w_xaxis.set_ticks([0.25,1.25,2.25])
ax.w_yaxis.set_ticks([0.25,1.25,2.25])
ax.w_xaxis.set_ticklabels(column_names,fontsize=fsize)
ax.w_yaxis.set_ticklabels(row_names,fontsize=fsize)
ax.set_zticks([0, 0.1,0.2,0.3,0.4])
ax.w_zaxis.set_ticklabels([r'$0$', r'$0.1$',r'$0.2$',r'$0.3$',r'$0.4$'],fontsize=fsize)
for t in ax.zaxis.get_major_ticks(): t.label.set_fontsize(fsize)
ax.set_xlabel(r'$a$',fontsize=fsize_label,labelpad=20)
ax.set_ylabel(r'$b$',fontsize=fsize_label,labelpad=20)
ax.zaxis.set_rotate_label(False)
ax.text(3.4, 0, 0.66, r'$p_{\tilde{a},\tilde{b}}(a,b)$', None,fontsize=fsize_label)
#ax.set_zlabel(r'$p_{\tilde{a},\tilde{b}}(a,b)$',fontsize=fsize_label,labelpad=20,
#              rotation=0)
#ax.set_title(r'$p_{\tilde{a},\tilde{b}}(a,b)$',fontsize=fsize_label,loc='right')
plt.xlim([0.05,2.5])
plt.ylim([0,2.45])
plt.show()
plt.tight_layout()
plt.savefig('plots/joint_pmf_blue.pdf',bbox_inches="tight")


fig = plt.figure(figsize=(12, 9))
ax = Axes3D(fig)

cs = ['gray', 'lightgray', 'white'] * ly

ax.bar3d(xpos,ypos,zpos, dx, dy, dz, color=cs, edgecolor='black',lw=3,shade=False)

#sh()
ax.w_xaxis.set_ticks([0.25,1.25,2.25])
ax.w_yaxis.set_ticks([0.25,1.25,2.25])
ax.w_xaxis.set_ticklabels(column_names,fontsize=fsize)
ax.w_yaxis.set_ticklabels(row_names,fontsize=fsize)
ax.set_zticks([0, 0.1,0.2,0.3,0.4])
ax.w_zaxis.set_ticklabels([r' $0$', r'$0.1$',r' $\;0.2$',r' $\;0.3$',r' $\;0.4$'])
for t in ax.zaxis.get_major_ticks(): t.label.set_fontsize(fsize_3D)
ax.set_xlabel(r'$a$',fontsize=fsize_label,labelpad=20)
ax.set_ylabel(r'$b$',fontsize=fsize_label,labelpad=20)
#ax.zaxis.set_rotate_label(False)
#ax.text(3.4, 0, 0.66, r'$p_{\tilde{a},\tilde{b}}(a,b)$', None,fontsize=fsize_label)
#ax.set_zlabel(r'$p_{\tilde{a},\tilde{b}}(a,b)$',fontsize=fsize_label,labelpad=20,
#              rotation=0)
#ax.set_title(r'$p_{\tilde{a},\tilde{b}}(a,b)$',fontsize=fsize_label,loc='right')
ax.text(-1, 1.3, 0.4, r'$p_{\tilde{a},\tilde{b}}(a,b)$', None,fontsize=fsize_label)
plt.xlim([0.05,2.5])
plt.ylim([0,2.45])
plt.tight_layout()
plt.savefig('plots/simple_example_3x3_joint_pmf.pdf',bbox_inches="tight")

fig = plt.figure(figsize=(12, 9))
cs = ['gray', 'lightgray', 'white']
x_ticks = [1, 2, 3]
marginal = [0.55,0.15,0.3]
plt.bar(x_ticks,marginal,width=0.5,color=cs, edgecolor='black',lw=3)
plt.xlim([0.5,3.5])
plt.ylim([0,0.7])
plt.xticks(x_ticks,fontsize=fsize)
plt.yticks(fontsize=fsize)
plt.xlabel(r'$a$',fontsize=fsize_label)
#plt.title(r'$p_{\tilde{a}}(a)$',fontsize=fsize_label,pad = 20)
#plt.ylabel(r'$p_{\tilde{a}}(a)$',fontsize=fsize_label,rotation=0,labelpad=45)
plt.savefig('plots/simple_example_3x3_marginal_pmf.pdf',bbox_inches="tight")

fig = plt.figure(figsize=(12, 9))
cs = ['white', 'white', 'white']
x_ticks = [1, 2, 3]
cond_unnorm = [0.05,0.05,0.2]
plt.bar(x_ticks,cond_unnorm,width=0.5,color=cs, edgecolor='black',lw=3)
plt.xlim([0.5,3.5])
plt.ylim([0,0.7])
plt.xticks(x_ticks,fontsize=fsize)
plt.yticks(fontsize=fsize)
plt.xlabel(r'$b$',fontsize=fsize_label)
#plt.title(r'$p_{\tilde{a},\tilde{b}}(3,b)$',fontsize=fsize_label,pad = 20)
# plt.ylabel(r'$p_{\tilde{a},\tilde{b}}(3,b)$',fontsize=fsize_label,rotation=0,labelpad=55)
plt.savefig('plots/simple_example_3x3_conditional_pmf_unnorm.pdf',bbox_inches="tight")

fig = plt.figure(figsize=(12, 9))
cs = ['white', 'white', 'white']
x_ticks = [1, 2, 3]
cond_norm = [0.16667,0.16667,0.6666]
plt.bar(x_ticks,cond_norm,width=0.5,color=cs, edgecolor='black',lw=3)
plt.xlim([0.5,3.5])
plt.ylim([0,0.7])
plt.xticks(x_ticks,fontsize=fsize)
plt.yticks(fontsize=fsize)
plt.xlabel(r'$b$',fontsize=fsize_label)
#plt.title(r'$p_{\tilde{b} | \tilde{a}}(b | 3)$',fontsize=fsize_label,pad = 20)
# plt.ylabel(r'$p_{\tilde{b} | \tilde{a}}(b | 3)$',fontsize=fsize_label,rotation=0,labelpad=55)
plt.savefig('plots/simple_example_3x3_conditional_pmf_norm.pdf',bbox_inches="tight")
```
