```python
import matplotlib.pyplot as plt
import numpy as np

n = 21
n_realizations = 3

T = np.array([[0.6, 0.1, 0.3],[ 0.2, 0.8, 0.3],[ 0.2 ,0.1, 0.4]])

p0 = [ np.array([1./3,1./3,1./3]), np.array([1,0,0]),np.array([0.1,0.2,0.7]) ]


for i in range(len(p0)):
    
    prob = np.zeros((3,n))
    prob[:,0] = p0[i]
    for j in range(n-1):
        prob[:,j+1] = np.dot(T,prob[:,j])
        
    plt.figure(figsize=(9, 6))  
    plt.xticks(np.arange(0,n,5),fontsize=24) 
    plt.yticks(fontsize=24) 
    plt.xlabel("i", fontsize=20)  
    plt.xlim( -0.5, n + 0.5 )
    plt.ylim( -0.1, 1.1 )
    plt.xlabel("Customer", fontsize=25)  
    plt.plot(np.arange(0,n),prob[0,:],color='black',linestyle="dashed",lw=2,marker='o',markersize=10,markerfacecolor="white",markeredgecolor="black", label='SF')
    plt.plot(np.arange(0,n),prob[1,:],color='black',linestyle="dashed",lw=2,marker='o',markersize=10,markerfacecolor="silver",markeredgecolor="black", label='LA')
    plt.plot(np.arange(0,n),prob[2,:],color='black',linestyle="dashed",lw=2,marker='o',markersize=10,markerfacecolor="black",markeredgecolor="black", label='SJ')
    plt.gcf().subplots_adjust(left=0.2)
    plt.gcf().subplots_adjust(bottom=0.15)
    plt.legend(fontsize=22)
    plt.show()
    plt.savefig('conv_markov_chain_' + str(i) + '.pdf',bbox_inches="tight")
 
```
