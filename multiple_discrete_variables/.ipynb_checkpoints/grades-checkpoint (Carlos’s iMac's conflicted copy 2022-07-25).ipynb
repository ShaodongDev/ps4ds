```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from scipy.stats import norm
import random
import time

math = pd.read_csv ("../data/student_grades/student-mat.csv", encoding='latin-1',sep=";")
port = pd.read_csv ("../data/student_grades/student-por.csv", encoding='latin-1',sep=";")

print(math.columns)

passed_math = math["G3"]>=10
print(passed_math)

font_size = 30
font_size_legend = 25
font_size_ticks = 25

print(np.sum(math["G3"]>=10)/len(math["G3"]))
print(np.sum(port["G3"]>=10)/len(port["G3"]))
```

```python
plt.figure()
plt.plot(math["G3"],math["age"],'.')

print(np.sum(passed_math[ math["romantic"]=="no" ]) / np.sum(math["romantic"]=="no"))
print(np.sum(passed_math[ math["romantic"]=="yes" ])/ np.sum(math["romantic"]=="yes"))

print(np.sum(passed_math[ math["address"]=="U" ]) / np.sum(math["address"]=="U"))
print(np.sum(passed_math[ math["address"]=="R" ])/ np.sum(math["address"]=="R"))

print(np.sum(passed_math[ math["age"]>16 ]) / np.sum(math["age"]>16))
print(np.sum(passed_math[ math["age"] <= 16 ])/ np.sum(math["age"]<=16))

print(np.sum((math["romantic"]=="no")[ math["age"]>16 ]) / np.sum(math["age"]>16))
print(np.sum((math["romantic"]=="yes")[ math["age"]>16 ]) / np.sum(math["age"]>16))

passed_older = passed_math[ math["age"]>16 ]
passed_younger = passed_math[ math["age"]<=16 ]

print(np.sum(passed_older[ math["romantic"]=="no" ]) / np.sum((math["romantic"]=="no")[ math["age"]>16 ]))
print(np.sum(passed_older[ math["romantic"]=="yes" ])/ np.sum((math["romantic"]=="yes")[ math["age"]>16 ]))

print(np.sum(passed_younger[ math["romantic"]=="no" ]) / np.sum((math["romantic"]=="no")[ math["age"]<=16 ]))
print(np.sum(passed_younger[ math["romantic"]=="yes" ])/ np.sum((math["romantic"]=="yes")[ math["age"]<=16 ]))


print(np.sum(math["age"]>16))
print(np.sum(math["age"]<=16))

print(np.sum(passed_math[ math["sex"]=="M" ]) / np.sum(math["sex"]=="M"))
print(np.sum(passed_math[ math["sex"]=="F" ])/ np.sum(math["sex"]=="F"))

print(np.sum(math["goout"] <= 2))
print(np.sum(math["goout"]>=3))

print(np.sum(passed_math[ math["goout"]<=2 ]) / np.sum(math["goout"]<=2))
print(np.sum(passed_math[ math["goout"]>=3 ]) / np.sum(math["goout"]>=3))

passed_goout_alot = passed_math[ math["goout"]>=3 ]
passed_goout_little = passed_math[ math["goout"]<=2 ]

print(np.sum((math["romantic"]=="no")[ math["goout"]>=3 ]))
print(np.sum((math["romantic"]=="no")[ math["goout"]<=2 ]))

print(np.sum((math["romantic"]=="yes")[ math["goout"]>=3 ]))
print(np.sum((math["romantic"]=="yes")[ math["goout"]<=2 ]))

print(np.sum(passed_goout_alot[ math["romantic"]=="no" ]) / np.sum((math["romantic"]=="no")[ math["goout"]>=3 ]))
print(np.sum(passed_goout_alot[ math["romantic"]=="yes" ]) / np.sum((math["romantic"]=="yes")[ math["goout"]>=3 ]))

print(np.sum(passed_goout_little[ math["romantic"]=="no" ])/ np.sum((math["romantic"]=="no")[ math["goout"]<=2 ]))
print(np.sum(passed_goout_little[ math["romantic"]=="yes" ])/ np.sum((math["romantic"]=="yes")[ math["goout"]<=2 ]))

print(np.sum((port["romantic"]=="no")[ port["school"] == "GP" ]))
print(np.sum((port["romantic"]=="yes")[ port["school"] == "GP" ]))
print(np.sum((port["romantic"]=="no")[ port["school"] == "MS" ]))
print(np.sum((port["romantic"]=="yes")[ port["school"] == "MS" ]))

print(np.sum( port["school"] == "GP"))
print(np.sum( port["school"] == "MS"))

print(np.sum( math["failures"] == 0))
print(np.sum( math["failures"] > 0))

print(np.sum(passed_math[ math["failures"]== 0 ]) / np.sum(math["failures"]==0))
print(np.sum(passed_math[ math["failures"]> 0 ]) / np.sum(math["failures"]>0))

passed_previous_failures = passed_math[ math["failures"] > 0 ]
passed_no_previous_failures = passed_math[ math["failures"] == 0 ]

print(np.sum(passed_previous_failures[ math["age"]>16]) / np.sum((math["age"]>16)[  math["failures"] > 0  ]))
print(np.sum(passed_previous_failures[ math["age"]<=16]) / np.sum((math["age"]<=16)[  math["failures"] > 0  ]))

print(np.sum(passed_no_previous_failures[ math["age"]>16]) / np.sum((math["age"]>16)[  math["failures"] == 0  ]))
print(np.sum(passed_no_previous_failures[ math["age"]<=16]) / np.sum((math["age"]<=16)[  math["failures"] == 0  ]))

print(np.sum( math["paid"] == "yes"))
print(np.sum( math["paid"] == "no"))

print(np.sum(passed_math[ math["paid"]== "yes" ]) / np.sum(math["paid"]=="yes"))
print(np.sum(passed_math[ math["paid"]== "no" ]) / np.sum(math["paid"]=="no"))

print(np.sum((math["paid"]=="yes")[ math["goout"]>=3 ]))
print(np.sum((math["paid"]=="yes")[ math["goout"]<=2 ]))

print(np.sum((math["paid"]=="no")[ math["goout"]>=3 ]))
print(np.sum((math["paid"]=="no")[ math["goout"]<=2 ]))

print(np.sum((math["paid"]=="yes")[ math["failures"] == 0 ]))
print(np.sum((math["paid"]=="yes")[ math["failures"] > 0 ]))

print(np.sum((math["paid"]=="no")[ math["failures"] == 0 ]))
print(np.sum((math["paid"]=="no")[ math["failures"] > 0 ]))

print(np.sum(passed_previous_failures[math["paid"]=="yes"]) / np.sum((math["paid"]=="yes")[  math["failures"] > 0  ]))
print(np.sum(passed_previous_failures[math["paid"]=="no"]) / np.sum((math["paid"]=="no")[  math["failures"] > 0  ]))

print(np.sum(passed_no_previous_failures[math["paid"]=="yes"]) / np.sum((math["paid"]=="yes")[  math["failures"] == 0  ]))
print(np.sum(passed_no_previous_failures[math["paid"]=="no"]) / np.sum((math["paid"]=="no")[  math["failures"] == 0  ]))
```
