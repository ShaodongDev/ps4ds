Code to simulate samples from a uniform distribution within a triangle\
Topics: Dependence between random variables, inverse-transform sampling

```python
from numpy import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
import numpy as np
import matplotlib
matplotlib.rcParams['text.usetex'] = True

rng = np.random.default_rng(2021)

font_size = 20
font_size_ticks = 15

n = 10000
u = rng.random((n,2))

markersize = 2
plt.figure(figsize=(4,4))
plt.scatter(u[:,1],u[:,0],s=markersize,color="black",rasterized=True)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylabel(r'$u_1$',fontsize=font_size,labelpad=5)
plt.xlabel(r'$u_2$',fontsize=font_size,labelpad=5)

a = 1-np.sqrt(1-u[:,0])

b_wrong = 1-np.sqrt(1-u[:,1])

plt.figure(figsize=(4,4))
plt.scatter(b_wrong,a,s=markersize,color="black",rasterized=True)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylabel(r'$a := 1 - \sqrt{1-u_1}$',fontsize=font_size,labelpad=5)
plt.xlabel(r'$b := 1 - \sqrt{1-u_2}$',fontsize=font_size,labelpad=5)
plt.title("Wrong simulation\n (assumes independence between variables)",fontsize=font_size)

b = (1-a) * u[:,1]

plt.figure(figsize=(4,4))
plt.scatter(b,a,s=markersize,color="black",rasterized=True)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylabel(r'$a := 1 - \sqrt{1-u_1}$',fontsize=font_size,labelpad=5)
plt.xlabel(r'$b := (1-a)u_2$',fontsize=font_size,labelpad=5)
plt.title("Correct simulation",fontsize=font_size)
```
