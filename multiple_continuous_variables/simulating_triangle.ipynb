Code to simulate samples from a uniform distribution within a triangle\
Topics: Dependence between random variables, inverse-transform sampling

```python
from numpy import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
import numpy as np

rng = np.random.default_rng(2021)

font_size = 15
font_size_ticks = 15

def generate_sample():
    u = rng.random(2)
    while u[1] > 2 * u[0] or u[1] > 2-2* u[0]:
        u = rng.random(2)
    return u

n = 10000
samples = np.zeros((n,2))
for ind in range(n):
    samples[ind,:] = generate_sample()

markersize = 1
plt.figure(figsize=(4,4))
plt.scatter(samples[:,0],samples[:,1],s=markersize,color="black")
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks);
```
