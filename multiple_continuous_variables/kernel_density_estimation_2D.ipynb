Simple example of kernel density estimation in 2D
Topics: Kernel density estimation, nonparametric density estimation

Author: Carlos Fernandez-Granda

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

font_size = 15
font_size_ticks = 15

xmin = -5
xmax = 5
ymin = -5
ymax = 5

x = np.linspace(xmin, xmax, 200)
y = np.linspace(ymin, ymax, 200)

def kernel(x,y,mu_x,mu_y,sigma):
    return (np.exp(-((x-mu_x)**2 + (y-mu_y)**2)/2/(sigma**2))
                  /(2 * np.pi * sigma))

x_grid, y_grid = np.meshgrid(x, y)

sigma = 1
pos_1 = [2,1,-1]
pos_2 = [2,1,-2]
z_1 = kernel(x_grid,y_grid,pos_1[0],pos_2[0],sigma)/3
z_2 = kernel(x_grid,y_grid,pos_1[1],pos_2[1],sigma)/3
z_3 = kernel(x_grid,y_grid,pos_1[2],pos_2[2],sigma)/3
kde = z_1 + z_2 + z_3

markersize = 60
plt.figure(figsize=(5,4))
plt.scatter(pos_1,pos_2,marker='x',s=markersize,color="black",linewidth=2)
plt.xticks(fontsize=font_size) 
plt.yticks(fontsize=font_size)
plt.ylim([ymin,ymax])
plt.xlim([xmin,xmax])
plt.ylabel('a[1]',fontsize=font_size,labelpad = 5)
plt.xlabel('a[2]',fontsize=font_size,labelpad = 5)
plt.show()
```

```python
zmin = 0
zmax = 0.08

fig = plt.figure(figsize=(5,4))
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(x_grid, y_grid, z_1, rstride=3, cstride=3, linewidth=1, antialiased=True,
                rasterized=True,cmap='Greys',vmin=0, vmax=0.08)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
ax.zaxis.set_tick_params(labelsize=font_size_ticks,pad=20)
plt.ylim([ymin,ymax])
plt.xlim([xmin,xmax])
ax.set_zlim3d(zmin,zmax)
plt.ylabel('a[2]',fontsize=font_size,labelpad = 5)
plt.xlabel('a[1]',fontsize=font_size,labelpad = 5)
plt.subplots_adjust(bottom=-0.01)

fig = plt.figure(figsize=(5,4))
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(x_grid, y_grid, z_2, rstride=3, cstride=3, linewidth=1, antialiased=True,
                rasterized=True,cmap='Greys',vmin=0, vmax=0.08)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
ax.zaxis.set_tick_params(labelsize=font_size_ticks,pad=20)
plt.ylim([ymin,ymax])
plt.xlim([xmin,xmax])
ax.set_zlim3d(zmin,zmax)
plt.ylabel('a[2]',fontsize=font_size,labelpad = 5)
plt.xlabel('a[1]',fontsize=font_size,labelpad = 5)
plt.subplots_adjust(bottom=-0.01)

fig = plt.figure(figsize=(5,4))
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(x_grid, y_grid, z_3, rstride=3, cstride=3, linewidth=1, antialiased=True,
                rasterized=True,cmap='Greys',vmin=0, vmax=0.08)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
ax.zaxis.set_tick_params(labelsize=font_size_ticks,pad=20)
plt.ylim([ymin,ymax])
plt.xlim([xmin,xmax])
ax.set_zlim3d(zmin,zmax)
plt.ylabel('a[2]',fontsize=font_size,labelpad = 5)
plt.xlabel('a[1]',fontsize=font_size,labelpad = 5)
plt.subplots_adjust(bottom=-0.01)

fig = plt.figure(figsize=(5,4))
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(x_grid, y_grid, kde, rstride=3, cstride=3, linewidth=1, antialiased=True,
                rasterized=True,cmap='Greys',vmin=0, vmax=0.08)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
ax.zaxis.set_tick_params(labelsize=font_size_ticks,pad=20)
plt.ylim([ymin,ymax])
plt.xlim([xmin,xmax])
ax.set_zlim3d(zmin,zmax)
plt.ylabel('a[2]',fontsize=font_size,labelpad = 5)
plt.xlabel('a[1]',fontsize=font_size,labelpad = 5)
```

```python
def fmt(x):
    s = f"{x:.4f}"
    if s.endswith("0"):
        s = f"{x:.2f}"
    return rf"{s}" if plt.rcParams["text.usetex"] else f"{s}"

plt.figure(figsize=(5,4))
CS = plt.contour(x_grid,y_grid, kde, levels=5,colors='black')
plt.clabel(CS, CS.levels, inline=True, fmt=fmt,fontsize=font_size-5)
plt.xticks(fontsize=font_size) 
plt.yticks(fontsize=font_size)
plt.ylim([ymin,ymax])
plt.xlim([xmin,xmax])
plt.xlabel('a[2]',fontsize=font_size,labelpad = 10)
plt.ylabel('a[1]',fontsize=font_size,labelpad = 15)
```
