```python
import os.path
import math
os.environ["PATH"] += os.pathsep + '/usr/local/texlive/2019/bin/x86_64-darwin'
print(os.getenv("PATH"))
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['text.usetex'] = True
from mpl_toolkits.mplot3d import Axes3D

font_size = 30
font_size_ticks = 22

x = np.linspace(-1.5, 1.5, 100)
y = np.linspace(-1.5, 1.5, 100)
invsq2 = 1 / np.sqrt(2)
U = np.array([[invsq2,invsq2],[invsq2,-invsq2]])
eigs = np.array([0.2,0.8])
D = np.diag(eigs)
Sigma = U @ D @ U.T
invSigma = np.linalg.inv(Sigma)

def gaussian(x,y):
    return (np.exp(-(invSigma[0,0] * x**2 + 2*invSigma[0,1]*x*y + invSigma[1,1] * y**2)/2)
                  /(2 * np.pi * np.sqrt(np.linalg.det(Sigma))))

X, Y = np.meshgrid(x, y)
Z = gaussian(X, Y)
c1=np.exp(-1/2)/(2 * np.pi * np.sqrt(np.linalg.det(Sigma)))
level_list = [c1]
level_list_2 = [1e-10,1e-7,1e-4,1e-2,1e-1,0.37]

fig = plt.figure(figsize = (9,9)) 
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z, rstride=3, cstride=3, linewidth=1, antialiased=True,
                cmap='Greys')#,vmin=0, vmax=0.08)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
ax.zaxis.set_tick_params(labelsize=font_size_ticks,pad=12)
#plt.ylim([ymin,ymax])
#plt.xlim([xmin,xmax])
#ax.set_zlim3d(zmin,zmax)
plt.ylabel('x[2]',fontsize=font_size,labelpad = 25)
plt.xlabel('x[1]',fontsize=font_size,labelpad = 25)
plt.savefig('plots/gaussian_3D.pdf',bbox_inches="tight")
plt.show()

fig = plt.figure(figsize = (9,9)) 
CS = plt.contour(X, Y, Z, levels = level_list,colors='black')#levels=30,cmap='RdGy',alpha=1);
plt.clabel(CS, fontsize=12, inline=1,fmt = '%1.2f')#,ticks=Contourrange)
CS2 = plt.contour(X, Y, Z, levels = level_list_2,colors='black')#levels=30,cmap='RdGy',alpha=1);
fmt2 = {}
strs = [r'$10^{-10}$',r'$10^{-7}$',r'$10^{-4}$',r'$10^{-2}$',r'$10^{-1}$','0.37']
for l, s in zip(CS2.levels, strs):
    fmt2[l] = s
plt.clabel(CS2, fontsize=14, inline=1,fmt=fmt2)# ['$10^{-10}$','$10^{-7}$','$10^{-5}$'])#,ticks=Contourrange)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.xlabel("x[1]", fontsize=font_size,labelpad=10)
h = plt.ylabel("x[2]", fontsize=font_size,labelpad=15)
h.set_rotation(0)
#plt.gcf().subplots_adjust(bottom=0.15)
plt.gcf().subplots_adjust(left=0.18)
plt.savefig('plots/gaussian_contours.pdf',bbox_inches="tight")

fig = plt.figure(figsize = (9,9)) 
CS = plt.contour(X, Y, Z, levels = level_list,colors='black',alpha=1)#levels=30,cmap='RdGy',alpha=1);
plt.clabel(CS, fontsize=12, inline=1,fmt = '%1.2f')#,ticks=Contourrange)
CS2 = plt.contour(X, Y, Z, levels = level_list_2,colors='black',alpha=0.2)#levels=30,cmap='RdGy',alpha=1);
plt.clabel(CS2, fontsize=14, inline=1,fmt=fmt2)# ['$10^{-10}$','$10^{-7}$','$10^{-5}$'])#,ticks=Contourrange)
ax = plt.axes()
ax.arrow(0, 0, invsq2*np.sqrt(eigs[0]), invsq2*np.sqrt(eigs[0]), head_width=0.1, 
         head_length=0.1, fc='black', ec='black',length_includes_head=True,lw=2)
ax.arrow(0, 0, invsq2*np.sqrt(eigs[1]), -invsq2*np.sqrt(eigs[1]), head_width=0.1, 
         head_length=0.1, fc='black', ec='black',length_includes_head=True,lw=2)
ax.text(invsq2*np.sqrt(eigs[0])-0.05, invsq2*np.sqrt(eigs[0])+0.05,r'$\mathbf{\sqrt{\lambda_2}\,u_2}$',
        fontsize=20)
ax.text(invsq2*np.sqrt(eigs[1])+0.03, -invsq2*np.sqrt(eigs[1])-0.15,r'$\mathbf{\sqrt{\lambda_1}\,u_1}$',
        fontsize=20)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.xlabel("x[1]", fontsize=font_size,labelpad=10)
h = plt.ylabel("x[2]", fontsize=font_size,labelpad=15)
h.set_rotation(0)
#plt.gcf().subplots_adjust(bottom=0.15)
plt.gcf().subplots_adjust(left=0.19)
plt.savefig('plots/gaussian_contours_arrows.pdf',bbox_inches="tight")

fig = plt.figure(figsize = (9,9)) 
CS = plt.contour(X, Y, Z, levels = level_list,colors='black',alpha=1)#levels=30,cmap='RdGy',alpha=1);
plt.clabel(CS, fontsize=12, inline=1,fmt = '%1.2f')#,ticks=Contourrange)
CS2 = plt.contour(X, Y, Z, levels = level_list_2,colors='black',alpha=0.2)#levels=30,cmap='RdGy',alpha=1);
plt.clabel(CS2, fontsize=14, inline=1,fmt=fmt2)# ['$10^{-10}$','$10^{-7}$','$10^{-5}$'])#,ticks=Contourrange)
ax = plt.axes()
ax.arrow(0, 0, invsq2*np.sqrt(eigs[0]), invsq2*np.sqrt(eigs[0]), head_width=0.1, 
         head_length=0.1, fc='black', ec='black',length_includes_head=True,lw=2)
ax.arrow(0, 0, invsq2*np.sqrt(eigs[1]), -invsq2*np.sqrt(eigs[1]), head_width=0.1, 
         head_length=0.1, fc='black', ec='black',length_includes_head=True,lw=2)
ax.text(invsq2*np.sqrt(eigs[0])-0.05, invsq2*np.sqrt(eigs[0])+0.05,r'$\mathbf{\sqrt{\lambda_2}\,u_2}$',
        fontsize=20)
ax.text(invsq2*np.sqrt(eigs[1])+0.03, -invsq2*np.sqrt(eigs[1])-0.15,r'$\mathbf{\sqrt{\lambda_1}\,u_1}$',
        fontsize=20)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.xlabel("x[1]", fontsize=font_size,labelpad=10)
h = plt.ylabel("x[2]", fontsize=font_size,labelpad=15)
h.set_rotation(0)
#plt.gcf().subplots_adjust(bottom=0.15)
plt.gcf().subplots_adjust(left=0.19)
plt.savefig('plots/gaussian_contours_arrows.pdf',bbox_inches="tight")
```
