```python
import matplotlib.pyplot as plt
import numpy as np
from os import listdir
import seaborn as sns
from scipy import stats

font_size = 25
font_size_ticks = 22

np.set_printoptions(precision=5)

def create_data_matrix(str_path):
    file_name_list = listdir(str_path)
    file_name_list.sort()
    col_ind = 8
    data_matrix = []
    ind = 0
    for file_name in file_name_list:
        if file_name[0] == '.':
            continue
        else:
            print("Station " + str(ind))
            print(file_name)
            ind = ind + 1
            data_aux = np.loadtxt(str_path + file_name, usecols=range(14))
            data = data_aux[:,col_ind]
            if len(data_matrix) == 0:
                data_matrix = data
            else:
                data_matrix = np.vstack((data_matrix,data))
    return data_matrix.T

def process_name(x):
    x = x[14:]
    x = x[:-7]
    x = x.translate(str.maketrans('','','_1234567890'))
    return x[2:] + ", " + x[:2]

str_path = "./data/weather/"

load_files = False
save_path = str_path + "hourly"
str_path_2015 = str_path + "hourly/2015/"
file_name_2015 = save_path + "hourly_temperatures_2015.npy"

if load_files:
    data_matrix = create_data_matrix(str_path_2015)
    np.save(file_name_2015,data_matrix_2015)
else:
    data_matrix = np.load(file_name_2015)

file_name_list = listdir(str_path + "hourly/2015/")
file_name_list.sort()

for ind,name in enumerate(file_name_list):
    print(ind)
    print(name)

print(data_matrix.shape)
print(data_matrix[25:35,:10])
```

```python
stations = [52,55,44] # [52,53] 55 Versailles
station_names = []
for station in stations:
    name = process_name(file_name_list[station+1])# list starts at 1
    station_names.append(name)
    print(name) 
```

```python
def remove_faulty_measurements(data,verbose):
    min_val = -100 # We assume temperatures cannot be lower than -100
    for ind in range(len(data)):
        median = np.median(data[data > min_val])
        if data[ind] < min_val:
            if verbose:
                print(str(ind) + ": " + str(data[ind]))
            aux_ind = ind-1
            while aux_ind > 0:
                if data[aux_ind] > min_val:
                    data[ind] = data[aux_ind]
                    print("changed to entry " + str(aux_ind) + ": " + str(data[ind]))
                    break
                else:
                    ind -= 1
    return data

station_1 = process_name(file_name_list[stations[0]+1])
station_2 = process_name(file_name_list[stations[1]+1])

verbose = True
temperature_1 = remove_faulty_measurements(data_matrix[:,stations[0]],verbose)
temperature_2 = remove_faulty_measurements(data_matrix[:,stations[1]],verbose)
temperature_3 = remove_faulty_measurements(data_matrix[:,stations[2]],verbose)
```

```python
def KDE_2D(data_1,data_2,bw=None):
    xmin = data_1.min()
    xmax = data_1.max()
    ymin = data_2.min()
    ymax = data_2.max()
    x,y = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
    grid = np.vstack([x.ravel(), y.ravel()])
    values = np.vstack([data_1, data_2])
    kernel = stats.gaussian_kde(values,bw_method=bw)
    kde = np.reshape(kernel(grid).T, x.shape)
    return x,y,kde

def fmt(x):
    s = f"{x:.4f}"
    if s.endswith("0"):
        s = f"{x:.0f}"
    return rf"{s}" if plt.rcParams["text.usetex"] else f"{s}"

xmin = -24
xmax = 40
ymin = -24
ymax = 40

markersize =1 
plt.figure(figsize=(9,9))
plt.scatter(temperature_1,temperature_2,s=markersize,color="black")
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([ymin,ymax])
plt.xlim([xmin,xmax])
plt.ylabel(station_names[1],fontsize=font_size,labelpad = 10)
plt.xlabel(station_names[0],fontsize=font_size,labelpad = 15)
plt.savefig('plots/temperature_scatterplot_manhattan_versailles.pdf',bbox_inches="tight")

bw_vals = [0.1,0.25,0.5]
levels = [2,5,10,15,20,25,30]
# levels_no_labels = [2.5,7.5,10,12.5,17.5,20,22.5]


for ind,bw in enumerate(bw_vals):
    x,y,kde = KDE_2D(temperature_1,temperature_2,bw=bw)
    plt.figure(figsize=(9,9))
    plt.scatter(temperature_1,temperature_2,s=1,color="gray",alpha=0.3)
    #CS = sns.kdeplot(data= np.column_stack((temperature_1,temperature_2)),color="black")
    #plt.contour(x, y, 1e4*kde, levels=levels_no_labels,colors='black')
    #CS = plt.contour(x, y, 1e4*kde, levels=levels_labels,colors='black')
    CS = plt.contour(x, y, 1e4*kde, levels=levels,colors='black')
    plt.clabel(CS, CS.levels, inline=True, fmt=fmt,fontsize=18)
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.ylim([ymin,ymax])
    plt.xlim([xmin,xmax])
    plt.ylabel(station_names[1],fontsize=font_size,labelpad = 10)
    plt.xlabel(station_names[0],fontsize=font_size,labelpad = 15)
    plt.savefig('plots/temperature_joint_pdf_manhattan_versailles_'+str(ind) +'.pdf',bbox_inches="tight")
```

```python
markersize =1 
plt.figure(figsize=(9,9))
plt.scatter(temperature_1,temperature_3,s=markersize,color="black")
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([ymin,ymax])
plt.xlim([xmin,xmax])
plt.ylabel(station_names[2],fontsize=font_size,labelpad = 20)
plt.xlabel(station_names[0],fontsize=font_size,labelpad = 15)
plt.savefig('plots/temperature_scatterplot_manhattan_hilo.pdf',bbox_inches="tight")

levels = [5,10,20,30,40,50]

for ind,bw in enumerate(bw_vals):
    x,y,kde = KDE_2D(temperature_1,temperature_3,bw=bw)
    plt.figure(figsize=(9,9))
    plt.scatter(temperature_1,temperature_3,s=1,color="gray",alpha=0.3)
    #CS = sns.kdeplot(data= np.column_stack((temperature_1,temperature_2)),color="black")
    #plt.contour(x, y, 1e4*kde, levels=levels_no_labels,colors='black')
    #CS = plt.contour(x, y, 1e4*kde, levels=levels_labels,colors='black')
    CS = plt.contour(x, y, 1e4*kde, levels=levels,colors='black')
    plt.clabel(CS, CS.levels, inline=True, fmt=fmt,fontsize=18)
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.ylim([ymin,ymax])
    plt.xlim([xmin,xmax])
    plt.ylabel(station_names[2],fontsize=font_size,labelpad = 20)
    plt.xlabel(station_names[0],fontsize=font_size,labelpad = 15)
    plt.savefig('plots/temperature_joint_pdf_manhattan_hilo_'+str(ind) +'.pdf',bbox_inches="tight")
```

```python
# Marginal distributions
x = np.linspace(xmin,xmax,100)
y_max_2 = 0.125
y_max_1 = 0.042

def KDE_1D(data,x,bw=None):
    kernel = stats.gaussian_kde(data,bw_method=bw)
    kde = kernel(x)
    return kde

for index,bw in enumerate(bw_vals):
    kde = KDE_1D(temperature_1,x,bw=bw)
    plt.figure(figsize=(9,9))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    #sns.kdeplot(temperature_1,bw=bw,color="black",lw=3,gridsize=1000, legend = False)#,color='white', edgecolor='black',linewidth=2)
    plt.plot(x,kde,color="black",lw=3)
    plt.xlim([xmin,xmax])
    plt.ylim([0,y_max_1])
    plt.ylabel(r"Marginal pdf $f_{kde}(t)$",fontsize=font_size,labelpad=20)
    plt.xlabel('Temperature in Manhattan, KS ($^\circ$C)',fontsize=font_size,labelpad=20)
    plt.savefig('plots/temperature_marginal_manhattan' + str(index) + '.pdf',bbox_inches="tight")
    
    kde = KDE_1D(temperature_2,x,bw=bw)
    plt.figure(figsize=(9,9))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.plot(x,kde,color="black",lw=3)
    #sns.kdeplot(temperature_2,bw=bw,color="black",lw=3,gridsize=1000, legend = False)#,color='white', edgecolor='black',linewidth=2)
    plt.xlim([xmin,xmax])
    plt.ylim([0,y_max_1])
    plt.ylabel(r"Marginal pdf $f_{kde}(t)$",fontsize=font_size,labelpad=20)
    plt.xlabel('Temperature in Versailles, KY ($^\circ$C)',fontsize=font_size,labelpad=20)
    plt.savefig('plots/temperature_marginal_versailles' + str(index) + '.pdf',bbox_inches="tight")
    
    kde = KDE_1D(temperature_3,x,bw=bw)
    plt.figure(figsize=(9,9))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.plot(x,kde,color="black",lw=3)
    #sns.kdeplot(temperature_3,bw=bw,color="black",lw=3,gridsize=1000, legend = False)#,color='white', edgecolor='black',linewidth=2)
    plt.xlim([xmin,xmax])
    plt.ylim([0,y_max_2])
    plt.ylabel(r"Marginal pdf $f_{kde}(t)$",fontsize=font_size,labelpad=20)
    plt.xlabel('Temperature in Hilo, HI ($^\circ$C)',fontsize=font_size,labelpad=20)
    plt.savefig('plots/temperature_marginal_hilo' + str(index) + '.pdf',bbox_inches="tight")
```

```python
# Conditional distributions computed from the 2D density estimate

def KDE_cond(data,data_cond,val_cond,t,bw_1D=None,bw_2D=None,verbose=False):
    xmin = data.min()
    xmax = data.max()
    x,y = np.meshgrid(t,val_cond)
    grid = np.vstack([x.ravel(), y.ravel()])
    if verbose:
        print("Grid:")
        print(grid)
    values = np.vstack([data, data_cond])
    kernel_2D = stats.gaussian_kde(values,bw_method=bw_2D)
    kde_2D = np.reshape(kernel_2D(grid).T, x.shape)
    if verbose:
        print("kde_2D: ")
        print(kde_2D)
    kernel_1D = stats.gaussian_kde(data_cond,bw_method=bw)
    kde_1D = kernel_1D(val_cond)
    if verbose:
        print("kde_1D: ")
        print(kde_1D)
    kde_cond = kde_2D / kde_1D
    return kde_cond[0]

bw = 0.25
verbose = False
t_cond = [-10,0,10,20,30]
y_max_1 = 0.16
y_max_2 = 0.185

kde = KDE_1D(temperature_2,x,bw=bw)
plt.figure(figsize=(9,9))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.plot(x,kde,color="black",lw=3)
#sns.kdeplot(temperature_2,bw=bw,color="black",lw=3,gridsize=1000, legend = False)#,color='white', edgecolor='black',linewidth=2)
plt.xlim([xmin,xmax])
plt.ylim([0,y_max_1])
plt.ylabel(r"Marginal pdf $f_{kde}(t)$",fontsize=font_size,labelpad=20)
plt.xlabel('Temperature in Versailles, KY ($^\circ$C)',fontsize=font_size,labelpad=20)
plt.savefig('plots/temperature_marginal_versailles_cond.pdf',bbox_inches="tight")

kde = KDE_1D(temperature_3,x,bw=bw)
plt.figure(figsize=(9,9))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.plot(x,kde,color="black",lw=3)
#sns.kdeplot(temperature_3,bw=bw,color="black",lw=3,gridsize=1000, legend = False)#,color='white', edgecolor='black',linewidth=2)
plt.xlim([xmin,xmax])
plt.ylim([0,y_max_2])
plt.ylabel(r"Marginal pdf $f_{kde}(t)$",fontsize=font_size,labelpad=20)
plt.xlabel('Temperature in Hilo, HI ($^\circ$C)',fontsize=font_size,labelpad=20)
plt.savefig('plots/temperature_marginal_hilo_cond.pdf',bbox_inches="tight")

for ind,t_val in enumerate(t_cond):
    
    x_2D,y_2D,kde = KDE_2D(temperature_1,temperature_2,bw=bw)
    plt.figure(figsize=(9,9))
    plt.scatter(temperature_1,temperature_2,s=1,color="gray",alpha=0.3)
    CS = plt.contour(x_2D, y_2D, 1e4*kde, levels=levels,colors='gray')
    plt.clabel(CS, CS.levels, inline=True, fmt=fmt,fontsize=18)
    plt.vlines(t_val,ymin,ymax,lw=3,ls="dashed")
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.ylim([ymin,ymax])
    plt.xlim([xmin,xmax])
    plt.ylabel(station_names[1],fontsize=font_size,labelpad = 20)
    plt.xlabel(station_names[0],fontsize=font_size,labelpad = 15)
    plt.savefig('plots/temperature_joint_pdf_manhattan_versailles_cond_'+str(ind) +'.pdf',bbox_inches="tight")
    
    kde = KDE_1D(temperature_1,x,bw=bw)
    plt.figure(figsize=(9,9))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.plot(x,kde,color="gray",lw=3)
    kernel_1D_manhattan = stats.gaussian_kde(temperature_1,bw_method=bw)
    plt.plot(t_val,kernel_1D_manhattan(t_val),'o',color="black",markersize=10)
    #sns.kdeplot(temperature_2,bw=bw,color="black",lw=3,gridsize=1000, legend = False)#,color='white', edgecolor='black',linewidth=2)
    plt.xlim([xmin,xmax])
    plt.ylim([0,y_max_1])
    plt.ylabel(r"Marginal pdf $f_{kde}(t)$",fontsize=font_size,labelpad=20)
    plt.xlabel('Temperature in Manhattan, KS ($^\circ$C)',fontsize=font_size,labelpad=20)
    plt.savefig('plots/temperature_marginal_manhattan_cond_'+str(ind)+'.pdf',bbox_inches="tight")
     
    kde = KDE_cond(temperature_2,temperature_1,t_val,x,bw_1D=bw,bw_2D=bw,verbose=verbose)
    plt.figure(figsize=(9,9))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.plot(x,kde,color="black",lw=3)
    plt.xlim([xmin,xmax])
    plt.ylim([0,y_max_1])
    plt.ylabel(r"Conditional pdf $f_{kde}(t|$"+str(t_val)+"$)$",fontsize=font_size,labelpad=20)
    plt.xlabel('Temperature in Versailles, KY ($^\circ$C)',fontsize=font_size,labelpad=20)
    plt.savefig('plots/temperature_conditional_versailles' + str(ind) + '.pdf',bbox_inches="tight")
    
    kde = KDE_cond(temperature_3,temperature_1,t_val,x,bw_1D=bw,bw_2D=bw,verbose=verbose)
    plt.figure(figsize=(9,9))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.plot(x,kde,color="black",lw=3)
    plt.xlim([xmin,xmax])
    plt.ylim([0,y_max_2])
    plt.ylabel(r"Conditional pdf $f_{kde}(t|$"+str(t_val)+"$)$",fontsize=font_size,labelpad=20)
    plt.xlabel('Temperature in Hilo, HI ($^\circ$C)',fontsize=font_size,labelpad=20)
    plt.savefig('plots/temperature_conditional_hilo' + str(ind) + '.pdf',bbox_inches="tight")
```

```python
# Conditional distributions computed directly
bw=0.25
epsilon = 0.5

for ind,t in enumerate(t_cond):
    
    indices = np.where(np.logical_and(temperature_1 >= t - epsilon, temperature_1 <= t + epsilon))
    temperature_cond_2 = temperature_2[indices]
    temperature_cond_3 = temperature_3[indices]
    
    print("Number of points, 2: " +str(len(temperature_cond_2)))
    print("Number of points, 3: " +str(len(temperature_cond_3)))
    
    markersize =1 
    plt.figure(figsize=(9,9))
    plt.scatter(temperature_1,temperature_2,s=markersize,color="gray",alpha=0.3)
    plt.scatter(temperature_1[indices],temperature_cond_2,s=markersize+1,color="black") 
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.ylim([ymin,ymax])
    plt.xlim([xmin,xmax])
    plt.ylabel(station_names[2],fontsize=font_size,labelpad = 20)
    plt.xlabel(station_names[0],fontsize=font_size,labelpad = 15)
    plt.savefig('plots/temperature_scatterplot_manhattan_versailles_cond_'+str(ind)+'.pdf',bbox_inches="tight")
    
    kde = KDE_1D(temperature_cond_2,x,bw=bw)
    plt.figure(figsize=(9,9))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.plot(x,kde,color="black",lw=3)
    sns.rugplot(temperature_cond_2,color="black")
    plt.xlim([xmin,xmax])
    plt.ylim([0,y_max_1])
    plt.ylabel(r"Conditional pdf $f_{kde}(t|$"+str(t_val)+"$)$",fontsize=font_size,labelpad=20)
    plt.xlabel('Temperature in Versailles, KY ($^\circ$C)',fontsize=font_size,labelpad=20)
    plt.savefig('plots/temperature_conditional_versailles' + str(ind) + '_emp.pdf',bbox_inches="tight")
    
    kde = KDE_1D(temperature_cond_3,x,bw=bw)
    plt.figure(figsize=(9,9))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.plot(x,kde,color="black",lw=3)
    sns.rugplot(temperature_cond_3,color="black")
    plt.xlim([xmin,xmax])
    plt.ylim([0,y_max_2])
    plt.ylabel(r"Conditional pdf $f_{kde}(t|$"+str(t_val)+"$)$",fontsize=font_size,labelpad=20)
    plt.xlabel('Temperature in Hilo, HI ($^\circ$C)',fontsize=font_size,labelpad=20)
    plt.savefig('plots/temperature_conditional_hilo' + str(ind) + '_emp.pdf',bbox_inches="tight")
```
