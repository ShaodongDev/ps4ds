```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from scipy.stats import norm
from scipy import stats

data = pd.read_csv ("../data/ANSUR II MALE Public.csv", encoding='latin-1')
print(data.columns[:20])
#print(data.columns[20:40])
#print(data.columns[40:70])
#print(data.columns[70:100])
#print(data.columns[100:140])

font_size = 25
font_size_legend = 22
font_size_ticks = 22
```

```python
stature = data['stature'] / 10.
head = data['headcircumference'] / 10
reported_height = data['Heightin'] * 2.54
weight = data['weightkg']/10
reported_weight = data['Weightlbs'] * 0.453592
foot = data['footlength']/10.

stature_min = stature.min()-10
stature_max = stature.max()+10
weight_min = weight.min()-10
weight_max = weight.max()+10
foot_min = foot.min()-2
foot_max = foot.max()+2

markersize =1 
plt.figure(figsize=(9,9))
plt.scatter(stature,weight,s=markersize,color="black")
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([weight_min,weight_max])
plt.xlim([stature_min,stature_max])
plt.ylabel('Height (cms)',fontsize=font_size,labelpad = 10)
plt.xlabel('Weight (kgs)',fontsize=font_size,labelpad = 15)
#plt.savefig('plots/temperature_scatterplot_manhattan_versailles.pdf',bbox_inches="tight")

plt.figure(figsize=(9,9))
plt.scatter(stature,foot,s=markersize,color="black")
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
```

```python
def KDE_2D(data_1,data_2,x,y,bw=None):
    grid = np.vstack([x.ravel(), y.ravel()])
    values = np.vstack([data_1, data_2])
    kernel = stats.gaussian_kde(values,bw_method=bw)
    kde = np.reshape(kernel(grid).T, x.shape)
    return kde

def fmt(x):
    s = f"{x:.4f}"
    if s.endswith("0"):
        s = f"{x:.0f}"
    return rf"{s}" if plt.rcParams["text.usetex"] else f"{s}"

stature_weight_x,stature_weight_y = np.mgrid[stature_min:stature_max:100j, weight_min:weight_max:100j]
stature_foot_x,stature_foot_y = np.mgrid[stature_min:stature_max:100j, foot_min:foot_max:100j]

bw_vals = [0.3,0.4,0.5]
levels_weight = [1,2,5,10,15,20,25,30]
levels_foot = [1,2,5,10,50,100,200,250,300]


for ind,bw in enumerate(bw_vals):
    kde = KDE_2D(stature,weight,stature_weight_x,stature_weight_y,bw=bw)
    plt.figure(figsize=(9,9))
    plt.scatter(stature,weight,s=markersize,color="gray")
    CS = plt.contour(stature_weight_x,stature_weight_y, 1e4*kde, levels=levels_weight,colors='black')
    plt.clabel(CS, CS.levels, inline=True, fmt=fmt,fontsize=18)
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.ylim([weight_min,weight_max])
    plt.xlim([stature_min,stature_max])
    plt.ylabel('Height (cms)',fontsize=font_size,labelpad = 10)
    plt.xlabel('Weight (kgs)',fontsize=font_size,labelpad = 15)
    plt.savefig('plots/us_army_gaussian_joint_pdf_stature_weight_'+str(ind) +'.pdf',bbox_inches="tight")
    
    kde = KDE_2D(stature,foot,stature_foot_x,stature_foot_y,bw=bw)
    plt.figure(figsize=(9,9))
    plt.scatter(stature,foot,s=markersize,color="gray")
    CS = plt.contour(stature_foot_x,stature_foot_y, 1e3*kde, levels=levels,colors='black')
    plt.clabel(CS, CS.levels, inline=True, fmt=fmt,fontsize=18)
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.ylim([foot_min,foot_max])
    plt.xlim([stature_min,stature_max])
    plt.ylabel('Height (cms)',fontsize=font_size,labelpad = 10)
    plt.xlabel('Foot length (cms)',fontsize=font_size,labelpad = 15)
    plt.savefig('plots/us_army_gaussian_joint_pdf_stature_foot_'+str(ind) +'.pdf',bbox_inches="tight")
```
