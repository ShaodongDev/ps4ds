```python
import os.path
import math
os.environ["PATH"] += os.pathsep + '/usr/local/texlive/2019/bin/x86_64-darwin'
print(os.getenv("PATH"))
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['text.usetex'] = True
from mpl_toolkits.mplot3d import Axes3D

font_size = 30
font_size_ticks = 25

xmin = -3
xmax = 3
ymin = -3
ymax = 3
x = np.linspace(xmin, xmax, 200)
y = np.linspace(ymin, ymax,200)

def gaussian(x,y,rho):
    return (np.exp(-( x**2 + 2*rho*x*y + y**2)/2/(1-rho**2))
                  /(2 * np.pi * np.sqrt(1-rho**2)))

def gaussian_cond(x,y,rho):
    return (np.exp(-( (y-rho * x)**2 )/2/(1-rho**2))
                  /(2 * np.pi * np.sqrt(1-rho**2)))

X, Y = np.meshgrid(x, y)

rho_vals = [0,0.5,0.9]
level_list = [1e-10,1e-7,1e-4,1e-2,1e-1,0.25]
cond_val = 1.0
f_max = 0.39

for ind,rho in enumerate(rho_vals):
    Z = gaussian(X, Y,rho)

    fig = plt.figure(figsize = (9,9)) 
    CS = plt.contour(X, Y, Z, levels = level_list,colors='gray')#levels=30,cmap='RdGy',alpha=1);
    plt.clabel(CS, fontsize=font_size, inline=1,fmt = '%1.2f')#,ticks=Contourrange)
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.vlines(cond_val,ymin,ymax,lw=3,ls="dashed")
    plt.xlabel("a", fontsize=font_size,labelpad=10)
    h = plt.ylabel("b", fontsize=font_size,labelpad=15)
    h.set_rotation(0)
    plt.ylim([ymin,ymax])
    plt.xlim([xmin,xmax])
    #plt.gcf().subplots_adjust(bottom=0.15)
    #plt.gcf().subplots_adjust(left=0.18)
    plt.savefig('plots/gaussian_conditionals_contours_'+str(ind)+'.pdf',bbox_inches="tight")
    plt.show()

    fig = plt.figure(figsize = (9,9)) 
    ax = fig.gca(projection='3d')
    z_cond = gaussian(cond_val, y,rho)
    ax.plot_surface(X, Y, Z, rstride=1, cstride=3, cmap='Greys',alpha=1,edgecolor='none')
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    ax.zaxis.set_tick_params(labelsize=font_size_ticks,pad=12)
    plt.ylim([ymin,ymax])
    plt.xlim([xmin,xmax])
    #ax.set_zlim3d(zmin,zmax)
    plt.ylabel('b',fontsize=font_size,labelpad = 25)
    plt.xlabel('a',fontsize=font_size,labelpad = 25)
    plt.savefig('plots/gaussian_conditional_3D_'+str(ind)+'.pdf',bbox_inches="tight")
    
    fig = plt.figure(figsize = (9,9)) 
    ax = fig.gca(projection='3d')
    z_cond = gaussian(cond_val, y,rho)
    ax.plot_surface(X, Y, Z, rstride=1, cstride=3, cmap='Greys',alpha=0.2,edgecolor='none')
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    ax.zaxis.set_tick_params(labelsize=font_size_ticks,pad=12)
    ax.plot(cond_val*np.ones(len(x)), y, z_cond,lw=3,color="black",alpha=1)
    plt.ylim([ymin,ymax])
    plt.xlim([xmin,xmax])
    #ax.set_zlim3d(zmin,zmax)
    plt.ylabel('b',fontsize=font_size,labelpad = 25)
    plt.xlabel('a',fontsize=font_size,labelpad = 25)
    plt.savefig('plots/gaussian_conditional_3D_'+str(ind)+'_slice.pdf',bbox_inches="tight")
    
    plt.figure(figsize=(9,9))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.plot(y,gaussian_cond(cond_val,y,rho),color="black",lw=3)
    plt.xlim([xmin,xmax])
    plt.ylim([0,f_max])
    plt.xlabel("b", fontsize=font_size,labelpad=10)
    plt.ylabel(r"Conditional density $f_{\tilde{b}|\tilde{a}}(b|1)$", fontsize=font_size,labelpad=15)
    plt.savefig('plots/gaussian_conditional_' + str(ind) + '.pdf',bbox_inches="tight")
    
```

```python
fig = plt.figure(figsize = (9,9)) 

def round_gaussian(x,y,var):
    return (np.exp(-( x**2 + y**2)/2/var)
                  /(2 * np.pi * var))
var = 0.5**2
Z_round = round_gaussian(X, Y,var)

ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z_round, rstride=1, cstride=3, cmap='Greys',edgecolor='none')#,vmin=0, vmax=0.08)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
ax.zaxis.set_tick_params(labelsize=font_size_ticks,pad=12)
#plt.ylim([ymin,ymax])
#plt.xlim([xmin,xmax])
#ax.set_zlim3d(zmin,zmax)
plt.ylabel('x[2]',fontsize=font_size,labelpad = 25)
plt.xlabel('x[1]',fontsize=font_size,labelpad = 25)
plt.savefig('plots/gaussian_3D_conditional_full.pdf',bbox_inches="tight")
plt.show()

x_cond = 0.25
y_cond = -1
#X_cond, Y_cond = np.meshgrid(x, y_cond)
z_cond_x = round_gaussian(x_cond, y,var)
z_cond_y = round_gaussian(x, y_cond,var)

eps = 0

fig = plt.figure(figsize = (9,9)) 
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z_round, rstride=1, cstride=3, cmap='Greys',alpha=0.15,edgecolor='none')#,vmin=0, vmax=0.08)
#ax.plot(x, y_cond*np.ones(len(x)), z_cond+eps,lw=3,color="black")
ax.plot(x_cond*np.ones(len(x)), y, z_cond_x+eps,lw=3,color="black",alpha=1)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
ax.zaxis.set_tick_params(labelsize=font_size_ticks,pad=12)
#plt.ylim([ymin,ymax])
#plt.xlim([xmin,xmax])
#ax.set_zlim3d(zmin,zmax)
plt.ylabel('x[2]',fontsize=font_size,labelpad = 25)
plt.xlabel('x[1]',fontsize=font_size,labelpad = 25)
plt.savefig('plots/gaussian_3D_conditional_1.pdf',bbox_inches="tight")
plt.show()

x_cond = 0.5
y_cond = -1
#X_cond, Y_cond = np.meshgrid(x, y_cond)
z_cond_x = round_gaussian(x_cond, y,var)
z_cond_y = round_gaussian(x, y_cond,var)
fig = plt.figure(figsize = (9,9)) 
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z_round, rstride=1, cstride=3, cmap='Greys',alpha=0.15,edgecolor='none')#,vmin=0, vmax=0.08)
#ax.plot(x, y_cond*np.ones(len(x)), z_cond+eps,lw=3,color="black")
ax.plot(x_cond*np.ones(len(x)), y, z_cond_x+eps,lw=3,color="black",alpha=1)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
ax.zaxis.set_tick_params(labelsize=font_size_ticks,pad=12)
#plt.ylim([ymin,ymax])
#plt.xlim([xmin,xmax])
#ax.set_zlim3d(zmin,zmax)
plt.ylabel('x[2]',fontsize=font_size,labelpad = 25)
plt.xlabel('x[1]',fontsize=font_size,labelpad = 25)
plt.savefig('plots/gaussian_3D_conditional_2.pdf',bbox_inches="tight")
plt.show()
```
