```python
import numpy as np
from sklearn.datasets import fetch_olivetti_faces
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['text.usetex'] = True
np.set_printoptions(precision=2)

font_size = 30
font_size_legend = 30
font_size_ticks = 25
font_size_graph = 30
```

```python
data = fetch_olivetti_faces()
identities = data.target
faces = data.images.reshape((len(data.images), -1))
print(faces.shape)
image_shape=(64,64)
```

```python
mean_face = np.mean(faces, axis=0)
print(mean_face.shape)

plt.figure()
plt.imshow(mean_face.reshape(image_shape), cmap=plt.cm.gray, #cmap=plt.cm.gray_r,
               interpolation='nearest',
               vmin=mean_face.min(), vmax=mean_face.max())
plt.xticks(())
plt.yticks(())
plt.savefig('plots/mean_face.pdf',bbox_inches='tight')
```

```python
sample_covariance_matrix = np.cov(faces.T)
eigvals,U = np.linalg.eig(sample_covariance_matrix)
U = U.astype(float)
```

```python
fig = plt.figure(figsize = (12,9)) 
plt.plot(eigvals[:399], markersize=10, color="white",  marker='o', markeredgewidth=2,
            linestyle = 'None',markeredgecolor="black")
#plt.xlim(-10,400)
#plt.ylim(0,21)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks([0,5,10,15,20],fontsize=font_size_ticks)
plt.xlabel(r"i", fontsize=font_size,labelpad=25)
plt.ylabel(r"Eigenvalue $\lambda_i$", fontsize=font_size,labelpad=10)
plt.savefig('plots/faces_eigvals.pdf',bbox_inches="tight")  


inds = [0,1,2,3,4,9,14,19,29,39,49,99,149,199,249,299,349,399]

for ind in inds:
    pd = U[:,ind]
    plt.figure()
    plt.imshow(pd.reshape(image_shape), cmap=plt.cm.gray, #cmap=plt.cm.gray_r,
                   interpolation='nearest',
                   vmin=pd.min(), vmax=pd.max())
    plt.xticks(())
    plt.yticks(())
    plt.savefig('plots/faces_pd_'+str(ind+1)+'.pdf',bbox_inches='tight')
    print(float(eigvals[ind]))
```

```python
selected_ind = 390
face = faces[selected_ind,:]

plt.figure()
plt.imshow(face.reshape(image_shape), cmap=plt.cm.gray, #cmap=plt.cm.gray_r,
               interpolation='nearest',
               vmin=face.min(), vmax=face.max())
plt.xticks(())
plt.yticks(())
plt.savefig('plots/face_'+str(selected_ind)+'.pdf',bbox_inches='tight')

centered_face = face - mean_face

k = 8

for ind in range(k):
    pd = U[:,ind]
    print(centered_face @ pd)
```

```python
k_vals = [5,10,20,30,50,100,150,200,300,350,399]

for k in k_vals:
    
    approximation = mean_face + centered_face @ U[:,:k] @ U[:,:k].T
    
    plt.figure()
    plt.imshow(approximation.reshape(image_shape), cmap=plt.cm.gray, #cmap=plt.cm.gray_r,
                   interpolation='nearest',
                   vmin=approximation.min(), vmax=approximation.max())
    plt.xticks(())
    plt.yticks(())
    plt.savefig('plots/faces_'+str(selected_ind)+'_approx_'+ str(k)+'.pdf',bbox_inches='tight')
    print(float(eigvals[ind])) 
```
