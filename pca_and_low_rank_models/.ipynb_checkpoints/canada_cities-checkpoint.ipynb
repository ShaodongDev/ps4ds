```python
import os.path
import math
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.rcParams['text.usetex'] = True

from matplotlib import ticker, cm
import numpy as np
import csv

filename = '../data/canada.csv'# '../../data/pca/egypt.csv'
with open(filename, newline='') as csvfile:
    dataset = list(csv.reader(csvfile))

font_size = 40
font_size_legend = 40
font_size_ticks = 35
font_size_graph = 40

n = len(dataset)-1
latitude = np.zeros(n)
longitude = np.zeros(n)

for i in range(n):
    latitude[i] = float(dataset[i+1][1])    
    longitude[i] = float(dataset[i+1][2])   
    
latitude_c = latitude - np.mean(latitude)
longitude_c = longitude - np.mean(longitude)
inds = np.arange(-np.pi,np.pi,1e-2)
```

```python
fig = plt.figure(figsize = (18,9)) 
plt.scatter(longitude,latitude, s=40, c="gray", marker='o', edgecolor="black")
#plt.scatter(np.mean(longitude),np.mean(latitude), s=400, c="black", marker='*', edgecolor="black",label="Sample mean")
plt.xlim(-145,-45)
plt.ylim(35,85)
plt.xticks(fontsize=font_size) 
plt.yticks(fontsize=font_size)
plt.ylabel("Latitude", fontsize=font_size+5,labelpad=10)
plt.xlabel("Longitude", fontsize=font_size+5,labelpad=10)
#plt.gcf().subplots_adjust(bottom=0.25)
#plt.gcf().subplots_adjust(left=0.15)
plt.savefig('plots/canada.pdf',bbox_inches="tight")

fig = plt.figure(figsize = (18,9)) 
plt.scatter(longitude,latitude, s=40, c="lightgray", marker='o', edgecolor="gray")
plt.scatter(np.mean(longitude),np.mean(latitude), s=900, c="black", marker='*', edgecolor="black",label="Sample mean")
plt.xlim(-145,-45)
plt.ylim(35,85)
plt.xticks(fontsize=font_size) 
plt.yticks(fontsize=font_size)
plt.ylabel("Latitude", fontsize=font_size+5,labelpad=10)
plt.xlabel("Longitude", fontsize=font_size+5,labelpad=10)
#plt.gcf().subplots_adjust(bottom=0.25)
#plt.gcf().subplots_adjust(left=0.15)
plt.legend(fontsize=font_size+5,loc = "upper left")
plt.savefig('plots/canada_mean.pdf',bbox_inches="tight")
print("Mean longitude:",np.round(np.mean(longitude),2),"Mean latitude:",np.round(np.mean(latitude),2))
```

```python
fig = plt.figure(figsize = (6,3)) 
plt.scatter(longitude_c,latitude_c, s=6, c="black", marker='o', edgecolor="black")
plt.xlim(-55,45)
plt.ylim(-15,35)
plt.xticks(fontsize=16) 
plt.yticks(fontsize=16)
plt.ylabel("Centered latitude", fontsize=16,labelpad=10)
plt.xlabel("Centered longitude", fontsize=16,labelpad=10)
# plt.xlim( -1, 30)
plt.gcf().subplots_adjust(bottom=0.25)
plt.gcf().subplots_adjust(left=0.2)
plt.savefig('plots/scatterplot_centered.pdf',bbox_inches="tight")
```

```python
X = np.array((longitude_c,latitude_c))
n = X.shape[1]
cov = (X @ X.T) / n
eigval,U = np.linalg.eig(cov)
u_1 = U[:,0]
u_2 = U[:,1]

print("First principal direction",u_1)
print("Variance in first principal direction",eigval[0])
print("Second principal direction",u_1)
print("Variance in second principal direction",eigval[0])
```

```python
line_list = [np.array([np.cos(np.pi/4),np.sin(np.pi/4)]),u_1,u_2]
x_axis = [r"Component in the direction of $b$",r"First principal component",
          r"Second principal component"]
labels = [r"Direction of $b$",r"Direction of $u_1$",
          r"Direction of $u_2$"]
for i,v in enumerate(line_list):
    fig = plt.figure(figsize = (9,9)) 
    plt.scatter(longitude_c,latitude_c, s=10, c="gray", marker='o')
    # ax = plt.axes()
    # ax.arrow(0, 0, np.cos(np.pi/4), np.sin(np.pi/4), head_width=0.1, head_length=0.1, fc='red', ec='red',length_includes_head=True)
    plt.plot( [100 * v[0],-100 * v[0]], [100*v[1], -100*v[1]],'--',color='black',lw=4,label=labels[i])
    plt.xlim(-55,45)
    plt.ylim(-45,55)
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.ylabel("Centered latitude", fontsize=font_size,labelpad=10)
    plt.xlabel("Centered longitude", fontsize=font_size,labelpad=10)
    # plt.xlim( -1, 30)
    #plt.gcf().subplots_adjust(bottom=0.15)
    #plt.gcf().subplots_adjust(left=0.15)
    plt.legend(fontsize=font_size,framealpha=1)
    plt.savefig('plots/scatterplot_proj_' + str(i) +'.pdf',bbox_inches="tight")
    
    proj_v_X = X.T @ v
    print('Variance of components: ' + str(np.var(proj_v_X)))
    print('Standard deviation of components: ' + str(np.std(proj_v_X)))
    
    fig = plt.figure(figsize = (9,9)) 
    plt.xlim(-55,45)
    plt.ylim(0,0.099)
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.ylabel("Density", fontsize=font_size,labelpad=10)
    plt.xlabel(x_axis[i], fontsize=font_size,labelpad=10)
    sns.distplot(proj_v_X, rug=True, hist=False, rug_kws={"color": "black"},
       kde_kws={"color": "black", "lw": 4,'bw_adjust': 0.6})
    #plt.gcf().subplots_adjust(bottom=0.15)
    #plt.gcf().subplots_adjust(left=0.15)
    plt.savefig('plots/scatterplot_proj_kde_' + str(i) +'.pdf',bbox_inches="tight")
```

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

def quad_form(x,y):
    return cov[0,0] * x**2 + 2*cov[0,1]*x*y + cov[1,1] * y**2

X, Y = np.meshgrid(x, y)
Z = quad_form(X, Y)
level_list = [10,100,250,500,1000,2000,3000]

tick_aux = np.arange(-2,3,1)

fig = plt.figure(figsize = (9,9)) 
CS = plt.contour(X, Y, Z, levels = level_list,colors='gray')#levels=30,cmap='RdGy');
plt.clabel(CS, fontsize=font_size_ticks, inline=1,fmt = '%1.0f')#,ticks=Contourrange)
plt.plot(np.cos(inds),np.sin(inds),'--',color='black',lw=3,label=r"$||a||_2=1$")
#axes = plt.axes()
#axes.arrow(0, 0, np.cos(np.pi/4), np.sin(np.pi/4), head_width=0.1, head_length=0.1, fc='red', ec='red',length_includes_head=True,lw=2)
plt.xticks(tick_aux,fontsize=font_size_ticks) 
plt.yticks(tick_aux,fontsize=font_size_ticks)
h = plt.ylabel(r"$a[2]$", fontsize=font_size,labelpad=10)
h.set_rotation(0)
plt.xlabel(r"$a[1]$", fontsize=font_size,labelpad=10)
plt.legend(fontsize=font_size,framealpha=1,loc="upper left")
#plt.gcf().subplots_adjust(bottom=0.15)
#plt.gcf().subplots_adjust(left=0.15)
plt.savefig('plots/quadform_noarrow.pdf',bbox_inches="tight")
```

```python
fig = plt.figure(figsize = (9,9)) 
CS = plt.contour(X, Y, Z, levels = level_list,colors='gray')#levels=30,cmap='RdGy');
plt.clabel(CS, fontsize=font_size_ticks, inline=1,fmt = '%1.0f')#,ticks=Contourrange)
plt.plot(np.cos(inds),np.sin(inds),'--',color='black',lw=3,label=r"$||a||_2=1$")
plt.arrow(0, 0, np.cos(np.pi/4), np.sin(np.pi/4), head_width=0.1, head_length=0.1, fc='black', ec='black',length_includes_head=True,lw=2)
plt.text(0.25,0.4,r"$b$",fontsize=font_size)
plt.xticks(tick_aux,fontsize=font_size_ticks) 
plt.yticks(tick_aux,fontsize=font_size_ticks)
h = plt.ylabel(r"$a[2]$", fontsize=font_size,labelpad=10)
h.set_rotation(0)
plt.xlabel(r"$a[1]$", fontsize=font_size,labelpad=10)
plt.legend(fontsize=font_size,framealpha=1,loc="upper left")
#plt.gcf().subplots_adjust(bottom=0.15)
#plt.gcf().subplots_adjust(left=0.15)
plt.savefig('plots/quadform.pdf',bbox_inches="tight")

fig = plt.figure(figsize = (9,9)) 
CS = plt.contour(X, Y, Z, levels = level_list,colors='gray')#levels=30,cmap='RdGy');
plt.clabel(CS, fontsize=font_size_ticks, inline=1,fmt = '%1.0f')#,ticks=Contourrange)
plt.plot(np.cos(inds),np.sin(inds),'--',color='black',lw=3,label=r"$||a||_2=1$")
plt.arrow(0, 0, u_1[0], u_1[1], head_width=0.1, head_length=0.1, fc='black', ec='black',
          length_includes_head=True,lw=2)
plt.text(1.075,-0.15,r"$u_1$",fontsize=font_size)
plt.xticks(tick_aux,fontsize=font_size_ticks) 
plt.yticks(tick_aux,fontsize=font_size_ticks)
h = plt.ylabel(r"$a[2]$", fontsize=font_size,labelpad=10)
h.set_rotation(0)
plt.xlabel(r"$a[1]$", fontsize=font_size,labelpad=10)
plt.legend(fontsize=font_size,framealpha=1,loc="upper left")
#plt.gcf().subplots_adjust(bottom=0.15)
#plt.gcf().subplots_adjust(left=0.15)
plt.savefig('plots/quadform_u1.pdf',bbox_inches="tight")

fig = plt.figure(figsize = (9,9)) 
CS = plt.contour(X, Y, Z, levels = level_list,colors='gray')#levels=30,cmap='RdGy');
plt.clabel(CS, fontsize=font_size_ticks, inline=1,fmt = '%1.0f')#,ticks=Contourrange)
plt.plot(np.cos(inds),np.sin(inds),'--',color='black',lw=3,label=r"$||a||_2=1$")
plt.arrow(0, 0, u_2[0], u_2[1], head_width=0.1, head_length=0.1, fc='black', ec='black',length_includes_head=True,lw=2)
plt.text(-0.25,0.6,r"$u_2$",fontsize=font_size)
plt.xticks(tick_aux,fontsize=font_size_ticks) 
plt.yticks(tick_aux,fontsize=font_size_ticks)
h = plt.ylabel(r"$a[2]$", fontsize=font_size,labelpad=10)
h.set_rotation(0)
plt.xlabel(r"$a[1]$", fontsize=font_size,labelpad=10)
plt.legend(fontsize=font_size,framealpha=1,loc="upper left")
#plt.gcf().subplots_adjust(bottom=0.15)
#plt.gcf().subplots_adjust(left=0.15)
plt.savefig('plots/quadform_u2.pdf',bbox_inches="tight")
```

```python
vecs = [U[:,1],-U[:,1]]
step = np.pi/16
mini_step = np.pi/64
for angle in np.arange(-np.pi/4-3*step,np.pi/4+3*step,step):
    vecs.append(np.array([np.cos(angle), np.sin(angle)]))
    vecs.append(np.array([-np.cos(angle), -np.sin(angle)]))
for angle in np.arange(np.pi/4+2.5*step+mini_step,np.pi/4+4*step+2*mini_step,mini_step):
    vecs.append(np.array([np.cos(angle), np.sin(angle)]))
    vecs.append(np.array([-np.cos(angle), -np.sin(angle)]))

fig = plt.figure(figsize = (9,9)) 
CS = plt.contour(X, Y, Z, levels = level_list,colors='gray',alpha=1)#levels=30,cmap='RdGy');
plt.clabel(CS, fontsize=font_size_ticks, inline=1,fmt = '%1.0f')#,ticks=Contourrange)
plt.plot(np.cos(inds),np.sin(inds),'--',color='black')
#ax = plt.axes()
for i,v in enumerate(vecs):
    Av = 2 * cov @ v
    Avn = Av / np.linalg.norm(Av)
    plt.arrow(v[0], v[1], Avn[0], Avn[1], head_width=0.1, head_length=0.1,lw=2, fc='black', ec='black',length_includes_head=True)
# plt.colorbar()
plt.xticks(tick_aux,fontsize=font_size_ticks) 
plt.yticks(tick_aux,fontsize=font_size_ticks)
h = plt.ylabel(r"$a[2]$", fontsize=font_size)
h.set_rotation(0)
plt.xlabel(r"$a[1]$", fontsize=font_size)
#plt.gcf().subplots_adjust(bottom=0.15)
#plt.gcf().subplots_adjust(left=0.15)
plt.savefig('plots/quadform_grad.pdf',bbox_inches="tight")
```
