Simulation of an exponential random variable using inverse-transform sampling\
Topics: Inverse-transform sampling

Author: Carlos Fernandez-Granda

```python
from numpy import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
import numpy as np

rng = np.random.default_rng(2021)

font_size = 15
font_size_legend = 15
font_size_ticks = 15

n = 10000
u = rng.random(n)
n_bins = 50

plt.figure(figsize=(4,3))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.hist(u,bins=n_bins,color='white', edgecolor='black', linewidth=2,
         density=True)
plt.title("Uniform samples",fontsize=font_size)

transformed_u = np.log(1/(1-u))

step = 1e-2
t=np.arange(0,np.max(transformed_u)+step,step)
exponential_pdf = expon.pdf(t,0,1)

y_max = 1.01

plt.figure(figsize=(4,3))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.hist(transformed_u,bins=n_bins,color='white', edgecolor='black', linewidth=2,
         density=True)
plt.ylim([0,y_max])
plt.title("Transformed samples",fontsize=font_size)

plt.figure(figsize=(4,3))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.hist(transformed_u,bins=n_bins,color='white', edgecolor='dimgray', linewidth=2,
         density=True)
plt.plot(t,exponential_pdf,lw=3,color="black",
     label=r'Exponential pdf')
plt.ylim([0,y_max])
plt.legend(fontsize=font_size)
plt.title("Comparison to exponential pdf",fontsize=font_size);
```
