```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
matplotlib.rcParams['text.usetex'] = True
from numpy.random import default_rng

rng = default_rng(2021)

font_size = 25
font_size_ticks = 20

vals = np.arange(1,6+1,1)
n = 60
true_pmf = [1/6.]*6
n_plots = 3
y_max_counts = 15

for ind in range(n_plots):
    counts = rng.multinomial(n,true_pmf)
    empirical_pmf = counts/n
    bin_edges = np.arange(0.5, 6.5+1, 1)
    plt.figure(figsize=(8,4))
    plt.bar(vals,empirical_pmf, width = 0.5, color = "lightgray", edgecolor="black",linewidth = 2)
    plt.xticks(np.arange(1, 6+1, 1))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.ylim([0,0.3])
    plt.ylabel(r"$p_{X}(a)$",fontsize=font_size,labelpad = 30)
    plt.xlabel(r"$a$",fontsize=font_size)
    
    plt.figure(figsize=(8,4))
    plt.bar(vals,counts, width = 0.5, color = "lightgray", edgecolor="black",linewidth = 2)
    plt.xticks(np.arange(1, 6+1, 1))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.ylim([0,y_max_counts])
    plt.ylabel(r"Counts",fontsize=font_size,labelpad = 30)
    plt.xlabel(r"Result",fontsize=font_size)
    
plt.figure(figsize=(8,4))
plt.bar(vals,true_pmf, width = 0.5, color = "lightgray", edgecolor="black",linewidth = 2)
plt.xticks(np.arange(1, 6+1, 1))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([0,0.3])
plt.ylabel(r"$p_{\tilde{a}}(a)$",fontsize=font_size,labelpad = 10)
plt.xlabel(r'$a$',fontsize=font_size)
```
