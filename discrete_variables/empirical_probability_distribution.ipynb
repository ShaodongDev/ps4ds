Analysis of the empirical probability estimator\
Topics: Binomial distribution, empirical probability

Author: Carlos Fernandez-Granda

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
matplotlib.rcParams['text.usetex'] = True
from scipy.stats import binom

theta = 0.5
n_vals = [20,40,80]
margin = 0.1

font_size = 25
font_size_legend = 25
font_size_ticks = 20

plt.figure(figsize=(12,6))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylabel(r"$p_{\tilde{\theta}}(t)$",fontsize=font_size,labelpad = 10)
plt.xlabel(r'$t$',fontsize=font_size)
markers = ['s','*','o']
for ind,n in enumerate(n_vals):
    
    possible_counts = np.arange(0,n+1,1)
    empirical_probabilities = possible_counts / n
    pmf = binom.pmf(possible_counts, n, theta)
    plt.plot(empirical_probabilities,pmf,markers[ind],markersize=10,markerfacecolor='white',markeredgecolor="black",label= 'n := ' + str(n))
    ind_1 = int(n * theta - n * margin)
    ind_2 = int(n * theta + n * margin)
    prob_error = np.sum(pmf[ind_1:(ind_2+1)])
    print("n: " +str(n))
    print("Probability that the error in the empirical probability estimate is smaller or equal to " + str(margin) + " : {:0.3f}".format(prob_error))

plt.legend(fontsize=font_size_legend)
plt.show()
```
