```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
matplotlib.rcParams['text.usetex'] = True
from scipy.stats import binom

theta = 0.5
n_vals = [20,50,100]

font_size = 25
font_size_legend = 20
font_size_ticks = 20

plt.figure(figsize=(9,6))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylabel(r"$p_{\tilde{a}}(a)$",fontsize=font_size,labelpad = 10)
plt.xlabel(r'$a$',fontsize=font_size)

for n in n_vals:
    
    possible_counts = np.arange(0,n+1,1)
    empirical_probabilities = possible_counts / n
    pmf = binom.pmf(possible_counts, n, theta)
    plt.plot(empirical_probabilities,pmf,'o',markersize=10,markerfacecolor='white',markeredgecolor="black",label= 'n = ' + str(n))
    
plt.legend(fontsize=font_size_legend)
plt.savefig('plots/empirical_probability_distribution.pdf',bbox_inches="tight")

```
