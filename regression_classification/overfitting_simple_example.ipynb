Cartoon illustration of generalization and overfitting in linear regression using a toy dataset with a single feature\

```python
import numpy as np
import matplotlib.pyplot as plt 
from numpy.random import default_rng
from sklearn.linear_model import LinearRegression
font_size = 15
font_size_ticks = 15

msize = 70
xmin = -2
xmax = 2
ymin = -3.25
ymax = 3.25
x_grid = np.linspace(xmin, xmax, 200)
sigma = 0.4
```

```python
n=50
rng = default_rng(2023)
x = rng.standard_normal(n)
x_test = rng.standard_normal(n)
true_beta = 1
true_alpha = 0
z = sigma * rng.standard_normal(n)
z_test = sigma * rng.standard_normal(n)
y = true_beta * x + true_alpha + z
y_test = true_beta * x_test + true_alpha + z_test

X = np.array([x]).T

linear_model = LinearRegression().fit(X, y)

xtick_aux = np.arange(-2,3,1)

plt.figure(figsize=(4,3))
plt.scatter(x,y,s=msize,color="white",edgecolor='black',lw=2, rasterized=True,label="Data")
plt.plot(x_grid,linear_model.coef_[0] * x_grid + linear_model.intercept_,lw=3,
         color="black",ls="solid",label="OLS model")
plt.xticks(xtick_aux,fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([ymin,ymax])
plt.xlim([xmin,xmax])
plt.xlabel('Feature',fontsize=font_size,labelpad = 10)
plt.ylabel('Response',fontsize=font_size,labelpad = 15)
plt.title('Training set',fontsize=font_size,pad = 5)

plt.figure(figsize=(4,3))
plt.scatter(x_test,y_test,s=msize,color="white",edgecolor='black',lw=2, rasterized=True,label="Data")
plt.plot(x_grid,linear_model.coef_[0] * x_grid + linear_model.intercept_,lw=3,
         color="black",ls="solid",label="OLS model")
plt.xticks(xtick_aux,fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([ymin,ymax])
plt.xlim([xmin,xmax])
plt.xlabel('Feature',fontsize=font_size,labelpad = 10)
plt.ylabel('Response',fontsize=font_size,labelpad = 15)
plt.title('Test set',fontsize=font_size,pad = 5);
```
