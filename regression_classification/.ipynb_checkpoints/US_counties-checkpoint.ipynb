```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
import datetime
import math
np.set_printoptions(precision=3)

font_size = 30
font_size_ticks = 25

data_all = pd.read_csv ("../data/county_health/analytic_data2019.csv")

data = data_all[data_all['County Ranked (Yes=1/No=0)']==1]

smoking = 100*data["Adult smoking raw value"].values.astype(float)
income = data['Median household income raw value'].values.astype(float)/1000
premature_mortality = data['Premature age-adjusted mortality raw value'].values.astype(float)
# premature_mortality = data['Premature death raw value'].values
# print(data["Premature age-adjusted mortality raw value"].isnull().values.any())
print(income.shape)
```

```python
print("Smoking")
print("Mean: " + str(np.mean(smoking)))
print("Standard deviation: " + str(np.std(smoking)))
print("Income")
print("Mean: " + str(np.mean(income)))
print("Standard deviation: " + str(np.std(income)))
print("Premature mortality")
print("Mean: " + str(np.mean(premature_mortality)))
print("Standard deviation: " + str(np.std(premature_mortality)))
print("Variance: " + str(np.var(premature_mortality)))
```

```python
print("Covariance between smoking and premature_mortality: " 
      + str(np.cov( smoking, premature_mortality)[0][1]))
print("Correlation between smoking and premature_mortality: " 
      + str(np.corrcoef( smoking, premature_mortality)[0][1]))
print("Covariance between income and premature_mortality: " 
      + str(np.cov( income, premature_mortality)[0][1]))
print("Correlation between income and premature_mortality: " 
      + str(np.corrcoef( income, premature_mortality)[0][1]))
```

```python
X = np.array([[smoking,income]])[0].T
print("Covariance between smoking and income: " 
      + str(np.cov( smoking, income)[0][1]))
print("Correlation between smoking and income: " 
      + str(np.corrcoef( smoking, income)[0][1]))
print("Covariance matrix of smoking and income: " 
      + str(np.cov( X.T)))
```

```python
# OLS coefficients
crosscovariance = np.array([np.cov( smoking, premature_mortality)[0][1],np.cov( income, premature_mortality)[0][1]])
coeff_OLS = np.linalg.inv(np.cov( X.T)) @ crosscovariance
intercept_OLS = np.mean(premature_mortality) - coeff_OLS.T @ np.array([np.mean(smoking),np.mean(income)])
print(coeff_OLS)
print(intercept_OLS)
```

```python
linear_model = LinearRegression().fit(X, premature_mortality)
print("Coefficients: " + str(linear_model.coef_))
print("Intercept: " + str(linear_model.intercept_))
print(linear_model.score(X,premature_mortality))
```

```python
print(linear_model.score(X,premature_mortality))

predictions = linear_model.predict(X)
errors = premature_mortality - predictions
variance_response = np.var(premature_mortality)
variance_predictions = np.var(predictions)
variance_errors = np.var(errors)
print("Variance of the response: ",variance_response)
print("Variance of the predictions: ",variance_predictions)
print("Variance of the errors: ",variance_errors)
print("R squared: ",variance_predictions/variance_response)
```

```python
standardized_smoking = (smoking - np.mean(smoking))/np.std(smoking)
standardized_income = (income - np.mean(income))/np.std(income)
standardized_premature_mortality = (premature_mortality - np.mean(premature_mortality))/np.std(premature_mortality)

X_standardized = np.array([[standardized_smoking,standardized_income]])[0].T
linear_model_standardized = LinearRegression(fit_intercept=True).fit(X_standardized, standardized_premature_mortality)
print(linear_model_standardized.coef_)
print(linear_model_standardized.intercept_)
print(linear_model_standardized.score(X_standardized,standardized_premature_mortality))
```

```python
standardized_features = np.array([standardized_smoking,standardized_income]).T
predictions = linear_model_standardized.predict(standardized_features)

variance_estimate = np.var(predictions)
variance_error = np.var(standardized_premature_mortality - predictions)

print("Variance of the estimate: " + str(variance_estimate))
print("Variance of the error: " + str(variance_error))

print("R2 for tobacco use: " + str(np.corrcoef( smoking, premature_mortality)[0][1]**2))
print("R2 for household income: " + str(np.corrcoef( income, premature_mortality)[0][1]**2))
```
