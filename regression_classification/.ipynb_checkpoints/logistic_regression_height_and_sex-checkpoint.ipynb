```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from scipy import stats
from scipy.stats import norm
from sklearn.linear_model import LogisticRegression

data_men = pd.read_csv ("../data/ANSUR II MALE Public.csv", encoding='latin-1')
data_women = pd.read_csv ("../data/ANSUR II FEMALE Public.csv", encoding='latin-1')

font_size = 30
font_size_legend = 30
font_size_ticks = 25

stature_men = data_men['stature'] / 10.
stature_women = data_women['stature'] / 10.
stature_all = np.concatenate((stature_men,stature_women))

print(len(stature_men))
print(len(stature_women))
print(len(stature_all))
```

```python
height = np.concatenate((stature_women,stature_men))
X = np.array([height])
n = len(height)
y = np.ones(n)
aux_ones = len(stature_women)
y[:aux_ones] = 0
print(X.shape)
logistic_model = LogisticRegression(solver='lbfgs',penalty='none').fit(X.T, y)
beta_ml = logistic_model.coef_[0][0]
alpha_ml = logistic_model.intercept_[0]
print("ML intercept:",alpha_ml)
print("ML coefficient",beta_ml)
```

```python
alpha_min = -95 #5 # 0.5 * alpha_ml 
alpha_max = -5 #95 # 2 * alpha_ml
beta_min = 0 #-0.8 # 2 * beta_ML
beta_max = 0.8 #0 #0.5 * beta_ML

def logistic_function(a):
    return 1/(1+np.exp(-a))
    
def log_likelihood(alpha, beta, x_0, x_1):
    llk = np.zeros(alpha.shape)
    for i in range(llk.shape[0]):
        for j in range(llk.shape[1]):
            llk[i,j] = ( np.sum( np.log( ( 1 - logistic_function(beta[i,j] * x_0 + alpha[i,j]) ) ) ) 
                        + np.sum( np.log(logistic_function(beta[i,j] * x_1 + alpha[i,j]))) )  
            # print(logistic_function(beta[i,j] * x_0 + alpha[i,j]))
            #print(np.log( ( 1 - logistic_function(beta[i,j] * x_0 + alpha[i,j]) ) ))
            #print(llk[i,j])
    return llk

alpha_mesh, beta_mesh = np.mgrid[alpha_min:alpha_max:100j,beta_min:beta_max:100j]
log_likelihood_vals = log_likelihood(alpha_mesh, beta_mesh, stature_men,stature_women)

log_likelihood_ML = log_likelihood(np.array([[alpha_ml]]), np.array([[beta_ml]]), stature_women,stature_men)
print("alpha_ml",alpha_ml)
print("beta_ml",beta_ml)
print("Maximum log likelihood",log_likelihood_ML)
```

```python
def fmt(x):
    s = f"{x:.1f}"
    if s.endswith("0"):
        s = f"{x:.0f}"
    return rf"{s}" if plt.rcParams["text.usetex"] else f"{s}"

alpha_1 = -15 
beta_1 = 0.1
alpha_2 = -90 
beta_2 = 0.5

level_list = [-2e5,-1.5e5,-1e5,-5e4,-5000]

plt.figure(figsize=(12,9))
ax = plt.axes()
# plt.axis("equal")
CS = plt.contour(alpha_mesh, beta_mesh, log_likelihood_vals, 50, levels=
                 level_list
                 ,colors='black')
plt.plot(alpha_1, beta_1, "d",markerfacecolor='black',markeredgecolor="black", lw=2,markersize=16, 
         # label=r'$\alpha_1 = 15, \beta_1 = -0.1$')
         label=r'$\alpha_1$, $\beta_1$')
plt.plot(alpha_2, beta_2, "s",markerfacecolor='black',markeredgecolor="black", lw=2,markersize=16, 
         # label=r'$\alpha_2 = 90, \beta_2 = -0.5$')
         label=r'$\alpha_2$, $\beta_2$')
plt.plot(alpha_ml, beta_ml, "o",markerfacecolor='black',markeredgecolor="black", lw=2,markersize=16, 
         #label=r'$\alpha_{ML} = 48.6, \beta_{ML} = -0.29$')
          label=r'$\alpha_{ML}$, $\beta_{ML}$')  
plt.clabel(CS, CS.levels, inline=True, fontsize=font_size_ticks,fmt=fmt)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks) 
plt.xlabel(r"$\alpha$", fontsize=font_size,labelpad=10,)  
plt.ylabel(r"$\beta$", fontsize=font_size,labelpad=20, rotation=0)  
plt.legend(fontsize=font_size,numpoints=1,framealpha=1,loc="upper right")
plt.savefig('plots/stature_logistic_reg_likelihood.pdf',bbox_inches="tight")

```
