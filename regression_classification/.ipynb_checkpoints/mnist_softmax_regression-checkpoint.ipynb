```python
import time
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import log_loss

font_size = 30
font_size_legend = 30
font_size_ticks = 25
```

```python
mnist = fetch_openml("mnist_784")
#     "mnist_784", version=1, return_X_y=True, as_frame=False, parser="pandas"
# )
```

```python
print(mnist.data.shape)
print(mnist.target.shape[0])
#data = mnist.data.to_numpy()
data = mnist.data
plt.figure()
plt.imshow(data[8,:].reshape(28,28), cmap=plt.cm.gray_r, 
interpolation='nearest')
print(mnist.target[8])
```

```python
np.random.seed(2023)
n_test = 35000
n_total = mnist.target.shape[0]
aux_ind = np.random.permutation(range(n_total))
ind_test = aux_ind[:n_test]
ind_train = aux_ind[n_test:]

data_train = data[ind_train,:]
labels_train = mnist.target[ind_train]
data_test = data[ind_test,:]
labels_test = mnist.target[ind_test]

print("Training set", data_train.shape,labels_train.shape)
print("Test set", data_test.shape,labels_test.shape)
```

```python
scaler = StandardScaler()
data_train = scaler.fit_transform(data_train)
data_test = scaler.transform(data_test)
```

```python
cov_matrix = np.cov(data_train.T,ddof=1)
eigvals,eigvecs = np.linalg.eig(cov_matrix)
plt.figure()
plt.semilogy(np.sort(eigvals))
```

```python
logistic_model = LogisticRegression(solver='lbfgs',penalty='none',multi_class='multinomial',tol=1e-4,max_iter=100000,
                                    verbose=1,warm_start=True).fit(data_train, labels_train)
```

```python
logistic_model.fit(data_train, labels_train)
```

```python
train_score = logistic_model.score(data_train, labels_train)
test_score = logistic_model.score(data_test, labels_test)
print("Training score",train_score)
print("Test score",test_score)

predicted_prob_train = logistic_model.predict_proba(data_train)
llk_train = log_loss(labels_train, predicted_prob_train)    
print("Training negative log-likelihood",llk_train)
predicted_prob_test = logistic_model.predict_proba(data_test)
llk_test = log_loss(labels_test, predicted_prob_test)    
print("Test negative log-likelihood",llk_test)
```

```python
logistic_model_reg = LogisticRegression(solver='lbfgs',penalty='l2',C=0.01,multi_class='multinomial',tol=1e-4,max_iter=100000,
                                    verbose=1).fit(data_train, labels_train)
```

```python
train_score = logistic_model_reg.score(data_train, labels_train)
test_score = logistic_model_reg.score(data_test, labels_test)
print("Training score",train_score)
print("Test score",test_score)

predicted_prob_train_reg = logistic_model_reg.predict_proba(data_train)
llk_train_reg = log_loss(labels_train, predicted_prob_train_reg)    
print("Training negative log-likelihood",llk_train_reg)
predicted_prob_test_reg = logistic_model_reg.predict_proba(data_test)
llk_test_reg = log_loss(labels_test, predicted_prob_test_reg)    
print("Test negative log-likelihood",llk_test_reg)
```

```python
coefs = logistic_model.coef_.copy()
coefs_max = coefs.max()
print("Max",coefs_max)
coefs_min = coefs.min()
print("Min",coefs_min)

aux_min = -20
aux_max = 20

for i in range(10):
    plt.figure(figsize=(10, 5))
    im = plt.imshow(
        coefs[i].reshape(28, 28),
        interpolation="nearest",
        cmap="gray",#plt.cm.RdBu
        vmin = aux_min,
        vmax = aux_max
    )
    plt.xticks(())
    plt.yticks(())
    cbar = plt.colorbar(im)
    cbar.ax.tick_params(labelsize=font_size_ticks)
```

```python
coefs_reg = logistic_model_reg.coef_.copy()
coefs_max = coefs_reg.max()
print("Max",coefs_max)
coefs_min = coefs_reg.min()
print("Min",coefs_min)

aux_min = -0.2
aux_max = 0.2

for i in range(10):
    plt.figure(figsize=(10, 5))
    im = plt.imshow(
        coefs_reg[i].reshape(28, 28),
        interpolation="nearest",
        cmap="gray",#plt.cm.RdBu
        vmin = aux_min,
        vmax = aux_max
    )
    plt.xticks(())
    plt.yticks(())
    cbar = plt.colorbar(im)
    cbar.ax.tick_params(labelsize=font_size_ticks)
```

```python
v_min = 0
v_max = 5
prob_zeros = predicted_prob_test[:,0]
ind_sort_zeros = np.argsort(prob_zeros)

for ind in np.arange(1,5):
    actual_ind = ind_sort_zeros[-ind]
    plt.figure()
    plt.imshow(data_test[actual_ind,:].reshape(28,28), cmap=plt.cm.gray_r, 
    interpolation='nearest',vmin=v_min,vmax=v_max)
    plt.colorbar()
    print(labels_test[actual_ind])

prob_zeros = predicted_prob_test[:,0]
ind_sort = np.argsort(prob_zeros)

for ind in np.arange(1,5):
    actual_ind = ind_sort[-ind]
    plt.figure()
    plt.imshow(data_test[actual_ind,:].reshape(28,28), cmap=plt.cm.gray_r, 
    interpolation='nearest',vmin=v_min,vmax=v_max)
    plt.colorbar()
    print(labels_test[actual_ind])
```

```python
prob_zeros_reg = predicted_prob_test_reg[:,0]
ind_sort = np.argsort(prob_zeros_reg)

for ind in np.arange(1,5):
    actual_ind = ind_sort[-ind]
    print(prob_zeros_reg[actual_ind])
    plt.figure()
    plt.imshow(data_test[actual_ind,:].reshape(28,28), cmap=plt.cm.gray_r, 
    interpolation='nearest' ,vmin=v_min,vmax=v_max)
    print(labels_test[actual_ind])
```
