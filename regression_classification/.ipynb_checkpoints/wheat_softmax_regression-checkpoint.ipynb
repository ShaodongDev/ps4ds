```python
import os.path
import numpy as np
import math
import matplotlib.pyplot as plt
np.set_printoptions(precision=2)
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import log_loss

font_size = 30
font_size_legend = 30
font_size_ticks = 25
font_size_graph = 30

dataset = np.loadtxt("../data/seeds_dataset.txt")
data = dataset[:,:-1]
labels = dataset[:,-1]

print(data.shape)

# Center the columns 
centered_data = data - data.mean(axis=0)
print("Mean",data.mean(axis=0))
print("Standard deviation",data.std(axis=0))


# Normalize each column (divide by standard deviation)
standardized_data = (centered_data  / centered_data.std(axis=0)).T
print(standardized_data.shape)
```

```python
msize=5
medgewidth=1
for feature_1 in range(7):
    for feature_2 in range(7):
        fig = plt.figure(figsize = (4,4)) 
        plt.title(str(feature_1)+','+str(feature_2))
        plt.plot(standardized_data[feature_1,:][labels==1],standardized_data[feature_2,:][labels==1], markersize=msize, color="white", marker='s', markeredgewidth=medgewidth,
                    linestyle = 'None',markeredgecolor="black",label="Kama")
        plt.plot(standardized_data[feature_1,:][labels==2],standardized_data[feature_2,:][labels==2], markersize=msize, color="black", marker='o', markeredgewidth=medgewidth,
                     linestyle = 'None',markeredgecolor="black",label="Rosa")
        plt.plot(standardized_data[feature_1,:][labels==3],standardized_data[feature_2,:][labels==3], markersize=msize, c="black", marker='x', markeredgewidth=medgewidth,
                     linestyle = 'None',markeredgecolor="black",label="Canadian")
#plt.xlim(x_min,x_max)
#plt.ylim(y_min,y_max)
#plt.xticks(fontsize=font_size_ticks) 
#plt.yticks(fontsize=font_size_ticks)
#plt.xlabel("First principal component", fontsize=font_size,labelpad=25)
#plt.ylabel("Second principal component", fontsize=font_size,labelpad=10)
#plt.legend(fontsize=font_size_legend)
#plt.savefig('plots/wheat_first2pcs.pdf',bbox_inches="tight")  
```

```python
feature_1 = 0
feature_2 = 5
data_train = standardized_data[[feature_1,feature_2],:].T
print(data_train.shape)
print(labels.shape)
logistic_model = LogisticRegression(solver='lbfgs',penalty='none',multi_class='multinomial').fit(data_train, labels)
#logistic_model = LogisticRegression(solver='lbfgs',penalty='none',multi_class='multinomial',tol=1e-4,max_iter=100000,
#                                    verbose=1,warm_start=True).fit(data_train, labels_train)
```

```python
train_score = logistic_model.score(data_train, labels)
print("Training score:",train_score)
```

```python
coefs = logistic_model.coef_.copy()
print(coefs)
intercepts = logistic_model.intercept_.copy()
print(intercepts)
```

```python
max_feature_1 = np.max(data_train[:,0]) + 0.1
max_feature_2 = np.max(data_train[:,1]) + 0.1
min_feature_1 = np.min(data_train[:,0]) - 0.1
min_feature_2 = np.min(data_train[:,1]) - 0.1

n_grid = 100
feature_1_aux = np.linspace(min_feature_1, max_feature_1, 100)  
feature_2_aux = np.linspace(min_feature_2, max_feature_2, 100) 

feature_1_grid, feature_2_grid = np.meshgrid(feature_1_aux, feature_2_aux)

for c in range(3):

    logit = (feature_1_grid * logistic_model.coef_[c][0] + feature_2_grid * logistic_model.coef_[c][1] 
             + logistic_model.intercept_[c] )
    
    plt.figure(figsize=(9,9))
    #plt.ylim([min_entorhinal,max_entorhinal])
    #plt.xlim([min_hippocampus,max_hippocampus])
    #plt.plot(hippocampus_norm_notAD_train,entorhinal_norm_notAD_train, 'o',markersize=markersize,markerfacecolor='lightgray',
    #         markeredgecolor="gray",label='Healthy')
    #plt.plot(hippocampus_norm_AD_train,entorhinal_norm_AD_train, 'o',markersize=markersize,markerfacecolor='black',
    #         markeredgecolor="black",label='Alzheimer\'s')
    CS = plt.contour(feature_1_aux, feature_2_aux,logit, levels=8,colors='black',linewidths=2,linestyles="dashed")
    #manual_locations = [(0.0007, 0.0002), (0.0003, 0.00043), (0.00025, 0.0004), (0.0002, 0.00035),(0.0002, 0.00025),(0.00015, 0.00018)]
    plt.clabel(CS, CS.levels, inline=True, fontsize=font_size_ticks,inline_spacing=10, rightside_up=True,fmt='%1.1f')#,manual=manual_locations)
    # plt.xlim([min_hippocampus,max_hippocampus])
    # plt.ylim([min_entorhinal,max_entorhinal])
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.xlabel("Area",fontsize=font_size,labelpad=20)
    plt.ylabel('Asymmetry coefficient',fontsize=font_size,labelpad=20)
    # plt.legend(fontsize=font_size,framealpha=1,loc="lower right")
    plt.savefig('plots/wheat_softmax_logits_'+str(c)+'.pdf',bbox_inches="tight")
```

```python
for c in range(3):

    dims_grid = feature_1_grid.shape
    print(dims_grid)
    probabilities = np.zeros(dims_grid)
    
    for ind_1 in range(dims_grid[1]):
        for ind_2 in range(dims_grid[1]):
            prob = logistic_model.predict_proba(
                np.array([feature_1_grid[ind_1,ind_2],feature_2_grid[ind_1,ind_2]]).reshape(1, -1))
            probabilities[ind_1,ind_2] = prob[0][c]
    
    levels_p = [0.01,0.1,0.25,0.5,0.75,0.9]
    plt.figure(figsize=(9,9))
    #plt.ylim([min_entorhinal,max_entorhinal])
    #plt.xlim([min_hippocampus,max_hippocampus])
#     plt.plot(hippocampus_norm_notAD_train,entorhinal_norm_notAD_train, 'o',markersize=markersize,markerfacecolor='lightgray',
#              markeredgecolor="gray",label='Healthy')
#     plt.plot(hippocampus_norm_AD_train,entorhinal_norm_AD_train, 'o',markersize=markersize,markerfacecolor='black',
#              markeredgecolor="black",label='Alzheimer\'s')
    CS = plt.contour(feature_1_aux, feature_2_aux,probabilities, levels=levels_p,colors='black',linewidths=2,linestyles="dashed")
    #manual_locations = [(0.55, 0.45), (0.35, 0.45), (0.25, 0.35), (0.25, 0.3),(0.35, 0.04),(0.25, 0.05)]
    plt.clabel(CS, CS.levels, inline=True, fontsize=font_size_ticks,inline_spacing=3, rightside_up=True,fmt='%1.2f')
              #,manual=manual_locations)
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.xlabel("Area",fontsize=font_size,labelpad=20)
    plt.ylabel('Asymmetry coefficient',fontsize=font_size,labelpad=20)
    #plt.xlim([min_hippocampus,max_hippocampus])
    #plt.ylim([min_entorhinal,max_entorhinal])
    #plt.legend(fontsize=font_size,framealpha=1,loc="lower right")
    #plt.savefig('plots/AD_logistic_reg_probability.pdf',bbox_inches="tight")

```
