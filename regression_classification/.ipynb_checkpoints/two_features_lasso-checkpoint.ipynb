```python
import numpy as np
import matplotlib.pyplot as plt 
from numpy.random import default_rng
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
font_size = 30
font_size_ticks = 25
np.set_printoptions(precision=3)

msize = 350
xmin = -3
xmax = 3
ymin = -3
ymax = 3
beta_x_min = -0.05
beta_x_max = 1.8
beta_y_min = -0.9
beta_y_max = 0.9
x_grid = np.linspace(xmin, xmax, 200)
sigma = 0.2
reps = 200
```

```python
rng = default_rng(2023)

n_vals = [8,80]
rho_sq_vals = np.sqrt([0.95,0.5])

for ind_n,n in enumerate(n_vals):

    for ind_rho,rho_sq in enumerate(rho_sq_vals): 
        common_component = rho_sq * rng.standard_normal(n)
        x_1_raw = common_component + np.sqrt(1-rho_sq**2) * rng.standard_normal(n)
        x_1 = x_1_raw - np.mean(x_1_raw)
        x_2_raw = common_component + np.sqrt(1-rho_sq**2) * rng.standard_normal(n)
        x_2 = x_2_raw - np.mean(x_2_raw)
        print(np.corrcoef(x_1,x_2))
        
        if ind_rho == 0 and ind_n == 0:
            x_1_ex = x_1
            x_2_ex = x_2
        true_beta_1 = 1
        true_beta_2 = 0
        true_alpha = 0

        plt.figure(figsize=(12,9))
        plt.scatter(x_1,x_2,s=msize,color="white",edgecolor='black',lw=2, rasterized=True)
        plt.xticks(fontsize=font_size_ticks) 
        plt.yticks(fontsize=font_size_ticks)
        plt.ylim([ymin,ymax])
        plt.xlim([xmin,xmax])
        plt.xlabel('Feature 1',fontsize=font_size,labelpad = 10)
        plt.ylabel('Feature 2',fontsize=font_size,labelpad = 15)
        #plt.legend(fontsize=font_size,framealpha=1)
        plt.savefig('plots/linear_regression_2_features_x_n'+str(ind_n)+'_rho'+str(ind_rho)+'.pdf',
                    bbox_inches="tight")
        
        betas = np.zeros((reps,2))

        for ind in range(reps):

            z = sigma * rng.standard_normal(n)
            y = true_beta_1 * x_1 + true_beta_2 * x_2 + true_alpha + z

            X = np.array([x_1,x_2]).T

            linear_model = LinearRegression(fit_intercept=False).fit(X, y)

            betas[ind,:] = linear_model.coef_

        plt.figure(figsize=(12,9))
        plt.scatter(betas[:,0],betas[:,1],s=msize/2,color="silver",edgecolor='silver',lw=2, rasterized=True)
        plt.scatter(true_beta_1,true_beta_2,s=3*msize,color="white",edgecolor='black',lw=3,marker="*", label="Ground truth")
        plt.xticks(fontsize=font_size_ticks) 
        plt.yticks(fontsize=font_size_ticks)
        plt.ylim([beta_min,beta_max])
        plt.xlim([beta_min,beta_max])
        plt.xlabel('Coefficient 1',fontsize=font_size,labelpad = 10)
        plt.ylabel('Coefficient 2',fontsize=font_size,labelpad = 15)
        plt.legend(fontsize=font_size,framealpha=1)
        plt.savefig('plots/linear_regression_2_features_betas_n'+str(ind_n)+'_rho'+str(ind_rho)+'.pdf',
                    bbox_inches="tight")
```

```python
rng = default_rng(2023)
n = 8

rho_sq = np.sqrt(0.95)
common_component = rho_sq * rng.standard_normal(n)
x_1_raw = common_component + np.sqrt(1-rho_sq**2) * rng.standard_normal(n)
x_1 = x_1_raw - np.mean(x_1_raw)
x_2_raw = common_component + np.sqrt(1-rho_sq**2) * rng.standard_normal(n)
x_2 = x_2_raw - np.mean(x_2_raw)
print(np.corrcoef(x_1,x_2))

true_beta_1 = 1
true_beta_2 = 0
true_alpha = 0

```

```python
rng = default_rng(2023)
n = 8
lambdas = [0.1,2,10]

betas = np.zeros((reps,2))

for ind_lambda,lambda_val in enumerate(lambdas):
    ridge_lambda = Ridge(alpha=lambda_val,fit_intercept=False)
        
    for ind in range(reps):

        z = sigma * rng.standard_normal(n)
        y = true_beta_1 * x_1 + true_beta_2 * x_2 + true_alpha + z

        X = np.array([x_1,x_2]).T

        linear_model = ridge_lambda.fit(X, y)

        betas[ind,:] = linear_model.coef_

    plt.figure(figsize=(12,9))
    plt.scatter(betas[:,0],betas[:,1],s=msize/2,color="silver",edgecolor='silver',lw=2, rasterized=True)
    plt.scatter(true_beta_1,true_beta_2,s=3*msize,color="white",edgecolor='black',lw=3,marker="*", label="Ground truth")
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.ylim([beta_min,beta_max])
    plt.xlim([beta_min,beta_max])
    plt.xlabel('Coefficient 1',fontsize=font_size,labelpad = 10)
    plt.ylabel('Coefficient 2',fontsize=font_size,labelpad = 15)
    plt.legend(fontsize=font_size,framealpha=1)
    plt.savefig('plots/linear_regression_2_features_ridge_lambda'+str(ind_lambda)+'.pdf',
                bbox_inches="tight")
```

```python
rng = default_rng(12345)
n = 8

x_1 = x_1_ex
x_2 = x_2_ex

z = sigma * rng.standard_normal(n)
y = true_beta_1 * x_1 + true_beta_2 * x_2 + true_alpha + z

beta_grid = np.linspace(beta_min, beta_max, 100)

def OLS(beta_1,beta_2):
    return (np.sum((y-x_1*beta_1-x_2*beta_2)**2))

def reg(beta_1,beta_2):
    return (beta_1**2 + beta_2**2)

def ridge(beta_1,beta_2,lambda_val):
    return (np.sum((y-x_1*beta_1-x_2*beta_2)**2) + lambda_val*(beta_1**2 + beta_2**2))

grid_1, grid_2 = np.meshgrid(beta_grid, beta_grid)

OLS_vec=np.vectorize(OLS)
grid_OLS = OLS_vec(grid_1,grid_2)

X = np.array([x_1,x_2]).T
linear_model = LinearRegression().fit(X, y)
beta_OLS = linear_model.coef_

level_list = [0.05,0.25,0.5,1,2,5,10,20]

fig = plt.figure(figsize = (9,9)) 
CS = plt.contour(grid_1, grid_2, grid_OLS, levels = level_list,colors='black')#levels=30,cmap='RdGy',alpha=1);
#CS2 = plt.contour(X, Y, Z_clean, levels = level_list_2,colors='black')#levels=30,cmap='RdGy',alpha=1);
plt.plot(true_beta_1,true_beta_2,ls="None",color='black',ms=20,marker="*",label="Ground truth")
plt.plot(beta_OLS[0],beta_OLS[1],ls="None",color='black',ms=12,marker="o",label="OLS estimate")
#ax = plt.axes()
#ax.text(1+0.2,1-0.1,r'$\mathbf{\beta_{OLS}}$',color='black',
#        fontsize=20)
plt.clabel(CS, fontsize=font_size_ticks, inline=1,fmt = '%1.2f')#,ticks=Contourrange)
#plt.clabel(CS2, fontsize=12, inline=1,fmt = '%1.2f')#,ticks=Contourrange)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.xlabel(r'$\beta[2]$', fontsize=font_size,labelpad=10)
h = plt.ylabel(r'$\beta [1] $', fontsize=font_size,labelpad=45)
h.set_rotation(0)
#plt.gcf().subplots_adjust(bottom=0.15)
#plt.gcf().subplots_adjust(left=0.18)
plt.legend(fontsize=font_size,framealpha=1,loc="lower right")
plt.savefig('plots/OLS_landscape.pdf',bbox_inches='tight')

level_list = [0.05,0.25,0.5,1,2,3,4,5,10,20]

grid_reg = reg(grid_1,grid_2)

fig = plt.figure(figsize = (9,9)) 
CS = plt.contour(grid_1, grid_2, grid_reg, levels = level_list,colors='black')#levels=30,cmap='RdGy',alpha=1);
#CS2 = plt.contour(X, Y, Z_clean, levels = level_list_2,colors='black')#levels=30,cmap='RdGy',alpha=1);
#plt.plot(true_beta_1,true_beta_2,ls="None",color='black',ms=20,marker="*",label="Ground truth")
#plt.plot(beta_OLS[0],beta_OLS[1],ls="None",color='black',ms=12,marker="o",label="OLS estimate")
#ax = plt.axes()
#ax.text(1+0.2,1-0.1,r'$\mathbf{\beta_{OLS}}$',color='black',
#        fontsize=20)
plt.clabel(CS, fontsize=font_size_ticks, inline=1,fmt = '%1.2f')#,ticks=Contourrange)
#plt.clabel(CS2, fontsize=12, inline=1,fmt = '%1.2f')#,ticks=Contourrange)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.xlabel(r'$\beta[2]$', fontsize=font_size,labelpad=10)
h = plt.ylabel(r'$\beta [1] $', fontsize=font_size,labelpad=45)
h.set_rotation(0)
#plt.gcf().subplots_adjust(bottom=0.15)
#plt.gcf().subplots_adjust(left=0.18)
#plt.legend(fontsize=font_size,framealpha=1,loc="lower right")
plt.savefig('plots/OLS_landscape.pdf',bbox_inches='tight')

level_list = [0.05,0.25,0.5,1,2,3,4,5,7,10,20]

ridge_vec=np.vectorize(ridge)
lambda_val = 2
grid_ridge = ridge_vec(grid_1,grid_2,lambda_val)

ridge_lambda = Ridge(alpha=lambda_val,fit_intercept=False)
ridge_model = ridge_lambda.fit(X, y)
beta_ridge = ridge_model.coef_

fig = plt.figure(figsize = (9,9)) 
CS = plt.contour(grid_1, grid_2, grid_ridge, levels = level_list,colors='black')#levels=30,cmap='RdGy',alpha=1);
#CS2 = plt.contour(X, Y, Z_clean, levels = level_list_2,colors='black')#levels=30,cmap='RdGy',alpha=1);
plt.plot(true_beta_1,true_beta_2,ls="None",color='black',ms=20,marker="*",label="Ground truth")
plt.plot(beta_OLS[0],beta_OLS[1],ls="None",color='black',ms=12,marker="o",label="OLS estimate")
plt.plot(beta_ridge[0],beta_ridge[1],ls="None",color='black',ms=16,marker="d",label="Ridge regression")
#ax = plt.axes()
#ax.text(1+0.2,1-0.1,r'$\mathbf{\beta_{OLS}}$',color='black',
#        fontsize=20)
plt.clabel(CS, fontsize=font_size_ticks, inline=1,fmt = '%1.2f')#,ticks=Contourrange)
#plt.clabel(CS2, fontsize=12, inline=1,fmt = '%1.2f')#,ticks=Contourrange)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.xlabel(r'$\beta[2]$', fontsize=font_size,labelpad=10)
h = plt.ylabel(r'$\beta [1] $', fontsize=font_size,labelpad=45)
h.set_rotation(0)
#plt.gcf().subplots_adjust(bottom=0.15)
#plt.gcf().subplots_adjust(left=0.18)
plt.legend(fontsize=font_size,framealpha=1,loc="lower right")
plt.savefig('plots/OLS_landscape.pdf',bbox_inches='tight')
```

```python
rng = default_rng(2023)
n = 8
lambdas = np.array([0.1,2,10])/n

x_1 = x_1_ex
x_2 = x_2_ex

betas = np.zeros((reps,2))

for ind_lambda,lambda_val in enumerate(lambdas):
    lasso_lambda = Lasso(alpha=lambda_val,fit_intercept=False)
        
    for ind in range(reps):

        z = sigma * rng.standard_normal(n)
        y = true_beta_1 * x_1 + true_beta_2 * x_2 + true_alpha + z

        X = np.array([x_1,x_2]).T

        linear_model = lasso_lambda.fit(X, y)

        betas[ind,:] = linear_model.coef_

    plt.figure(figsize=(12,9))
    plt.scatter(betas[:,0],betas[:,1],s=msize/2,color="silver",edgecolor='silver',lw=2, rasterized=True)
    plt.scatter(true_beta_1,true_beta_2,s=3*msize,color="white",edgecolor='black',lw=3,marker="*", label="Ground truth")
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.ylim([beta_min,beta_max])
    plt.xlim([beta_min,beta_max])
    plt.xlabel('Coefficient 1',fontsize=font_size,labelpad = 10)
    plt.ylabel('Coefficient 2',fontsize=font_size,labelpad = 15)
    plt.legend(fontsize=font_size,framealpha=1)
    plt.savefig('plots/linear_regression_2_features_lasso_lambda'+str(ind_lambda)+'.pdf',
                bbox_inches="tight")
```
