```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy import stats

font_size = 25
font_size_legend = 22
font_size_ticks = 22

data = pd.read_csv("./data/shot_logs.csv")
print(data.columns)
data_3s = data[data["PTS_TYPE"] == 3]
data_3s.head()
```

```python
minimum_3s = 100

taken = data_3s.groupby(['player_name'])['SHOT_RESULT'].count()
print(taken.head())
plt.figure()
plt.hist(taken)
print(len(taken[taken >= minimum_3s]))

selected_players = taken[taken >= minimum_3s].keys()
n_players = len(selected_players)
```

```python
unique_game_ids = np.sort(np.unique(data_3s["GAME_ID"].values))
unique_game_ids[:4]
n_games = len(unique_game_ids)
half = int(n_games/2)
half_id = unique_game_ids[half]
print(n_games)
first_half = unique_game_ids[:half]
second_half = unique_game_ids[half:]
```

```python
p_values = np.zeros(n_players)
percentages = np.zeros(n_players)
percentages_clutch = np.zeros(n_players)

for ind,player in enumerate(selected_players):#enumerate(["stephen curry"]):#enumerate(selected_players):
    print(player)
    player_data = data_3s[data_3s["player_name"] == player]
    player_data_firsthalf = data_3s["GAME_ID"]
    shots = player_data["SHOT_RESULT"].values
    attempted = len(shots)
    made = len(shots[shots == "made"])
    percentage = made / attempted
    percentages[ind] = percentage
    print(str(made) + " out of " + str(attempted)
          + " (" + str(np.round(percentage,3)) + ")" )
    data_fourth_quarter = player_data[player_data.PERIOD == 4]
    clutch_data = data_fourth_quarter[np.abs(data_fourth_quarter.FINAL_MARGIN)<10]
    clutch_shots = clutch_data["SHOT_RESULT"].values
    clutch_attempted = len(clutch_shots)
    clutch_made = len(clutch_shots[clutch_shots == "made"])
    if clutch_attempted > 0:
        percentage_clutch = clutch_made/clutch_attempted
        percentages_clutch[ind] = percentage_clutch
        print("CLUTCH: " + str(clutch_made) + " out of " + str(clutch_attempted)
              + " (" + str(np.round(percentage_clutch,3)) + ")" )
    # sorted_shots = df.sort_values(by=['GAME_ID'])
    # shot_counter = 0
        #pmf = binom.pmf(np.arange(0,clutch_made + 1), clutch_attempted, percentage)
        pmf = binom.pmf(np.arange(clutch_made, clutch_attempted+ 1), clutch_attempted, percentage)
        p_val = np.sum(pmf)
        print(np.arange(clutch_made, clutch_attempted+ 1))
        print("p value: " + str(np.round(p_val,4)))
        p_values[ind] = p_val 
    else:
        percentage_clutch = -1
```
