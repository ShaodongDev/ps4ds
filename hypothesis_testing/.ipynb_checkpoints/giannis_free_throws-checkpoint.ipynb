```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
matplotlib.rcParams['text.usetex'] = True

from numpy.random import default_rng
rng = default_rng(2021)

font_size = 25
font_size_legend = 22
font_size_ticks = 22

n = 100
p_null = 1./6

x = np.arange(0,n+1,1)
x_max = 35
pmf = binom.pmf(x, n, p_null)
cum_sum_pmf = np.cumsum(pmf)

plt.figure(figsize=(15,6))
plt.bar(x,pmf, width = 1, color = "white", edgecolor="black",linewidth = 2)
#plt.xticks(np.arange(1, 6+1, 1))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.xticks(np.arange(0,x_max+1, 5),fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.xlim([-0.5,x_max+0.5])
plt.ylabel(r"P(x 3s) under null hypothesis",fontsize=font_size,labelpad = 10)
plt.xlabel(r'$x$',fontsize=font_size)
plt.show()
plt.savefig('plots/test_die_pmf_null.pdf',bbox_inches="tight")

plt.figure(figsize=(15,6))
plt.bar(x,cum_sum_pmf, width = 1, color = "white", edgecolor="black",linewidth = 2,log=False)
plt.bar([13],[cum_sum_pmf[13]], width = 1, color = "gray", edgecolor="black",linewidth = 2,log=False,
        label="Observed 3s, p value = "+str(np.round(cum_sum_pmf[13],3)))
plt.hlines(0.05,-1,x_max+1,lw=3,ls="dashed",label="Significance threshold")
plt.xticks(np.arange(0,x_max+1, 2),fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.xlim([-0.5,x_max+0.5])
#plt.ylim([5e-6,1])
plt.ylabel(r"P(x 3s or less) under null hypothesis",fontsize=font_size,labelpad = 10)
plt.xlabel(r'x',fontsize=font_size)
plt.legend(fontsize=font_size)
plt.show()
plt.savefig('plots/test_die_p_value.pdf',bbox_inches="tight")
```
