```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy import stats

font_size = 15
font_size_legend = 15
font_size_ticks = 15

data = pd.read_csv("../data/p.values.one.per.paper.abstracts.csv")
pvalues = data['p.value'].values
dp_values, dp_counts = np.unique(data['decimal.places'], return_counts=True)
```

```python
rounded = np.round(pvalues,2)
rounded_values, rounded_counts = np.unique(rounded, return_counts=True)
print(rounded_values, rounded_counts)

plt.figure(figsize=(21,9))
plt.plot(rounded_values,rounded_counts,"o",markersize=15,markerfacecolor='black',markeredgecolor="black",
         markeredgewidth=2)
plt.ylabel(r"Counts",fontsize=font_size,labelpad = 20)
plt.xlabel(r'P values',fontsize=font_size,labelpad = 20)
plt.xticks(fontsize=font_size_ticks)
plt.yticks(fontsize=font_size_ticks)
plt.ylim([0,13000])
plt.grid(True)
```

```python
print("Two decimal places")
two_dp_data = data[data['decimal.places']==2]
two_dp_values, two_dp_counts = np.unique(two_dp_data['p.value'].values, return_counts=True)
print(two_dp_values,two_dp_counts)
```

```python
print("Three decimal places")
three_dp_data = data[data['decimal.places']==3]
three_dp_pvalues = three_dp_data['p.value'].values
three_dp_values, three_dp_counts = np.unique(three_dp_pvalues, return_counts=True)
print(three_dp_values, three_dp_counts)
three_dp_rounded = np.round(three_dp_pvalues,2)
three_dp_rounded_values, three_dp_rounded_counts = np.unique(three_dp_rounded, return_counts=True)
print(three_dp_rounded_values, three_dp_rounded_counts)
```

```python
print("Five decimal places")
five_dp_data = data[data['decimal.places']==5]
five_dp_pvalues = five_dp_data['p.value'].values
five_dp_values, five_dp_counts = np.unique(five_dp_pvalues, return_counts=True)
#print(five_dp_values, five_dp_counts)
five_dp_rounded = np.round(five_dp_pvalues,2)
five_dp_rounded_values, five_dp_rounded_counts = np.unique(five_dp_rounded, return_counts=True)
print(five_dp_rounded_values, five_dp_rounded_counts)
```

```python
rounded_1 = np.zeros(4)
rounded_1_vals = np.arange(0.01,0.05,0.01)
print(rounded_1_vals)
rounded_2_vals = np.arange(0.01,0.05,0.01)
unrounded = np.copy(pvalues)
for ind,vals in enumerate(rounded_1_vals):
    rounded_1[ind] = np.sum(pvalues == vals)
    unrounded = np.delete(unrounded, np.where(unrounded ==vals))    
unrounded_rounded = np.round(unrounded,2)
values, counts = np.unique(unrounded_rounded, return_counts=True)
print(rounded_1)
print(values)
print(counts)
plt.figure(figsize=(12,9))
plt.plot(values[1:5],counts[1:5]/np.sum(counts[1:5]),"o")
plt.plot(rounded_1_vals,rounded_1/np.sum(rounded_1),"x")
```

```python
rounded_1 = np.zeros(4)
rounded_2 = np.zeros(4)
rounded_1_vals = np.arange(0.01,0.05,0.01)
print(rounded_1_vals)
rounded_2_vals = np.arange(0.001,0.01,0.001)
print(rounded_2_vals)
unrounded = np.copy(pvalues)
unrounded = np.round(unrounded,5)
for ind,vals in enumerate(rounded_1_vals):
    rounded_1[ind] = np.sum(pvalues == vals)
    unrounded = np.delete(unrounded, np.where(unrounded ==vals))
aux_vals_1 = np.arange(0,0.05,0.01)
for ind,vals in enumerate(aux_vals_1):
    for ind_2,vals_2 in enumerate(rounded_2_vals): 
        val_2 = np.round(vals+vals_2,3)
        #print(val_2)
        if vals_2 < 0.005 and ind > 0:
            #print("assigned to " + str(rounded_1_vals[ind-1]))
            rounded_2[ind-1] += np.sum(pvalues == val_2)
        if vals_2 >= 0.005 and vals < 0.04:
            #print("assigned to " + str(rounded_1_vals[ind]))
            rounded_2[ind] += np.sum(pvalues == val_2)
        unrounded = np.delete(unrounded, np.where(unrounded == val_2))
values, counts = np.unique(unrounded, return_counts=True)
unrounded_rounded = np.round(unrounded,2)
values, counts = np.unique(unrounded_rounded, return_counts=True)
print(rounded_1)
print(np.sum(rounded_1))
print(rounded_2)
print(np.sum(rounded_2))
print(values)
print(counts)
print(np.sum(counts))

plt.figure(figsize=(11,9))
plt.plot(rounded_1_vals,rounded_1/np.sum(rounded_1),"x",markersize=10,markerfacecolor='black',markeredgecolor="black",
         markeredgewidth=3,label=r"$1$ significant figure")
plt.plot(rounded_1_vals,rounded_2/np.sum(rounded_2),"s",markersize=10,markerfacecolor='white',markeredgecolor="black",
         markeredgewidth=2,label=r"$2$ significant figures")
plt.plot(values[1:5],counts[1:5]/np.sum(counts[1:5]),"o",markersize=10,markerfacecolor='black',markeredgecolor="black",
         markeredgewidth=2,label=r"$\geq2$ significant figures")
plt.ylabel(r"Normalized counts",fontsize=font_size,labelpad = 20)
plt.xlabel(r'P values',fontsize=font_size,labelpad = 20)
plt.xticks(rounded_1_vals,fontsize=font_size_ticks)
plt.yticks(fontsize=font_size_ticks)
plt.xlim([0.005,0.045])
plt.legend(fontsize=font_size_ticks)
plt.savefig('plots/p_hacking_rounding.pdf',bbox_inches="tight")
plt.show()
```
