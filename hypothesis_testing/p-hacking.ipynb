Data source: https://journals.plos.org/plosbiology/article?id=10.1371/journal.pbio.1002106

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy import stats

font_size = 15
font_size_legend = 15
font_size_ticks = 15

data = pd.read_csv("../data/p.values.one.per.paper.abstracts.csv")
pvalues = data['p.value'].values
dp_values, dp_counts = np.unique(data['decimal.places'], return_counts=True)
```

```python
rounded = np.round(pvalues,2)
rounded_values, rounded_counts = np.unique(rounded, return_counts=True)
# print(rounded_values, rounded_counts)

plt.figure(figsize=(9,4))
plt.plot(rounded_values,rounded_counts,"o",markersize=8,markerfacecolor='black',markeredgecolor="black",
         markeredgewidth=2)
plt.ylabel(r"Counts",fontsize=font_size,labelpad = 5)
plt.xlabel(r'P values',fontsize=font_size,labelpad = 5)
plt.xticks(fontsize=font_size_ticks)
plt.yticks(fontsize=font_size_ticks)
plt.ylim([0,13000])
plt.grid(True);
```
