```python
hippo_min = ymin # hippocampus_norm_AD_train.min()
hippo_max = ymax # hippocampus_norm_AD_train.max()
ento_min = xmin # entorhinal_norm_AD_train.min()
ento_max = xmax # entorhinal_norm_AD_train.max()

ind_y,ind_x = np.mgrid[hippo_min:hippo_max:100j, ento_min:ento_max:100j]

def gaussian(x,y,mu,invSigma,Sigma):
    return (np.exp(-(invSigma[0,0] * (x-mu[0])**2 + 2*invSigma[0,1]*(x-mu[0])*(y-mu[1]) + invSigma[1,1] * (y-mu[1])**2)/2)
                  /(2 * np.pi * np.sqrt(np.linalg.det(Sigma))))

def fmt(x):
    s = f"{x:.4f}"
    if s.endswith("0"):
        s = f"{x:.1f}"
    return rf"{s}" if plt.rcParams["text.usetex"] else f"{s}"
```

```python
features_AD_train = np.vstack((hippocampus_norm_AD_train,entorhinal_norm_AD_train))
mu_AD = np.mean(features_AD_train,axis=1)
Sigma_AD = np.cov(features_AD_train)
print(mu_AD)
print(Sigma_AD)
invSigma_AD = np.linalg.inv(Sigma_AD)
z_gaussian_AD = gaussian(ind_x,ind_y,mu_AD,invSigma_AD,Sigma_AD)

features_notAD_train = np.vstack((hippocampus_norm_notAD_train,entorhinal_norm_notAD_train))
mu_notAD = np.mean(features_notAD_train,axis=1)
Sigma_notAD = np.cov(features_notAD_train)
print(mu_notAD)
print(Sigma_notAD)
invSigma_notAD = np.linalg.inv(Sigma_notAD)
z_gaussian_notAD = gaussian(ind_x,ind_y,mu_notAD,invSigma_notAD,Sigma_notAD)


levels_gaussian = [0.1,1,5]
alpha = 0.25

plt.figure(figsize=(12,9))
#plt.scatter(hippocampus_norm_AD_train,entorhinal_norm_AD_train,s=markersize,color="gray",alpha=alpha)
CS_notAD = plt.contour(ind_x,ind_y,z_gaussian_notAD/1e7, levels=levels_gaussian,colors='gray',linestyles='dashed',linewidths=3)
CS_notAD.collections[0].set_label("Healthy")
CS_AD = plt.contour(ind_x,ind_y,z_gaussian_AD/1e7, levels=levels_gaussian,colors='black',linewidths=3)
CS_AD.collections[0].set_label("Alzheimer's")
plt.clabel(CS_AD, CS_AD.levels, inline=True, fmt=fmt,fontsize=font_size_ticks)
plt.clabel(CS_notAD, CS_notAD.levels, inline=True, fmt=fmt,fontsize=font_size_ticks)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.xlim([ento_min,ento_max])
plt.ylim([hippo_min,hippo_max])
plt.ylabel(r"Hippocampus volume",fontsize=font_size,labelpad=20)
plt.xlabel('Entorhinal cortex volume',fontsize=font_size,labelpad=20)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
ax = plt.gca()
ax.yaxis.get_offset_text().set_fontsize(font_size_ticks-4)
ax.xaxis.get_offset_text().set_fontsize(font_size_ticks-4)
plt.legend(fontsize=font_size,framealpha=1,loc="lower right")
plt.savefig('plots/AD_QDA_gaussians.pdf',bbox_inches="tight")

plt.figure(figsize=(9,9))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.plot(hippocampus_norm_notAD_train,entorhinal_norm_notAD_train, 'o',markersize=markersize,markerfacecolor='lightgray',markeredgecolor="gray",
         label='Healthy',alpha=1)
plt.plot(hippocampus_norm_AD_train,entorhinal_norm_AD_train, 'o',markersize=markersize,markerfacecolor='black',markeredgecolor="black",
         label='Alzheimer\'s',alpha=1)
CS_AD = plt.contour(ind_x,ind_y,z_gaussian_AD, levels=8,colors='black',linewidths=3)
                 # 1e4*z_gaussian_stature_weight, levels=levels_weight,colors='black')
CS_notAD = plt.contour(ind_x,ind_y,z_gaussian_notAD, levels=8,colors='gray',linestyles='dashed',linewidths=3)
plt.xlim([0.0002,0.0007])
#plt.ylim([0.001,0.0035])
plt.ylabel(r"Hippocampus volume",fontsize=font_size,labelpad=20)
plt.xlabel('Entorhinal cortex volume',fontsize=font_size,labelpad=20)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
ax = plt.gca()
ax.yaxis.get_offset_text().set_fontsize(font_size_ticks-4)
ax.xaxis.get_offset_text().set_fontsize(font_size_ticks-4)


```

```python
ind_y,ind_x = np.mgrid[hippo_min:hippo_max:1000j, ento_min:ento_max:1000j]

p_AD = pd.value_counts(labels_train)["AD"] / len(labels_train)
def conditional_p_AD(x,y):
    p = p_AD * gaussian(x,y,mu_AD,invSigma_AD,Sigma_AD) / (
        p_AD * gaussian(x,y,mu_AD,invSigma_AD,Sigma_AD) +
        (1-p_AD) * gaussian(x,y,mu_notAD,invSigma_notAD,Sigma_notAD) )
    return p

cond_prob_AD = conditional_p_AD(ind_x,ind_y)

threshold = np.zeros(cond_prob_AD.shape[0])
aux_end = -1
flag = True
for ind in range(cond_prob_AD.shape[0]):
    idy = (np.abs(cond_prob_AD[ind,:] - 0.5)).argmin()
    threshold[ind] = ind_y[ind,idy]
    if idy == 0 and flag:
        aux_end = ind
        flag = False

plt.figure(figsize=(12,9))
plt.pcolormesh(ind_x,ind_y,cond_prob_AD, cmap='Greys',shading='auto' )
plt.plot(ind_x[:aux_end],threshold[:aux_end],ls="dashed",lw=2,color="black")
plt.colorbar()
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([ento_min,ento_max])
plt.xlim([hippo_min,hippo_max])

levels_p = [0.01,0.1,0.25,0.5,0.75]#np.arange(0.1,1,0.1)
plt.figure(figsize=(12,9))
plt.xlim([ento_min,ento_max])
plt.ylim([hippo_min,hippo_max])
plt.plot(hippocampus_norm_notAD_train,entorhinal_norm_notAD_train, 'o',markersize=markersize,markerfacecolor='lightgray',markeredgecolor="gray",
         label='Healthy',alpha=1)
plt.plot(hippocampus_norm_AD_train,entorhinal_norm_AD_train, 'o',markersize=markersize-1,markerfacecolor='black',markeredgecolor="black",
         label='Alzheimer\'s',alpha=1)
CS = plt.contour(ind_x,ind_y,cond_prob_AD, levels=levels_p,colors='black',linewidths=2,linestyles="dashed")
manual_locations = [(0.00015, 0.0004), (0.00015, 0.0003), (0.00015, 0.00025), (0.00015, 0.00015),(0.00015, 0.0001)]
plt.clabel(CS, CS.levels, inline=True, fontsize=font_size_ticks,inline_spacing=3, rightside_up=True,manual=manual_locations)
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylabel(r"Hippocampus volume",fontsize=font_size,labelpad=20)
plt.xlabel('Entorhinal cortex volume',fontsize=font_size,labelpad=20)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
ax = plt.gca()
ax.yaxis.get_offset_text().set_fontsize(font_size_ticks-4)
ax.xaxis.get_offset_text().set_fontsize(font_size_ticks-4)
plt.legend(fontsize=font_size,framealpha=1,loc="lower right")
plt.savefig('plots/AD_QDA_probability.pdf',bbox_inches="tight")
```

```python
def evaluate_QDA(hippocampus_norm_AD,entorhinal_norm_AD,hippocampus_norm_notAD,entorhinal_norm_notAD,threshold):
    
    AD_predicted_as_AD = 0
    AD_predicted_as_notAD = 0
    notAD_predicted_as_notAD = 0
    notAD_predicted_as_AD = 0


    for ind in range(len(hippocampus_norm_AD)):
        if conditional_p_AD(hippocampus_norm_AD[ind],entorhinal_norm_AD[ind]) > threshold:
            AD_predicted_as_AD += 1
        else:
            AD_predicted_as_notAD += 1
    for ind in range(len(hippocampus_norm_notAD)):
        if conditional_p_AD(hippocampus_norm_notAD[ind],entorhinal_norm_notAD[ind]) > threshold:
            notAD_predicted_as_AD += 1
        else:
            notAD_predicted_as_notAD += 1
    error_rate = 1- (AD_predicted_as_AD + notAD_predicted_as_notAD)/(
        AD_predicted_as_AD + notAD_predicted_as_notAD + AD_predicted_as_notAD 
        + notAD_predicted_as_AD)
    return [error_rate, AD_predicted_as_AD, AD_predicted_as_notAD,notAD_predicted_as_notAD,notAD_predicted_as_AD]


[error_rate_train, AD_predicted_as_AD_train, AD_predicted_as_notAD_train,
 notAD_predicted_as_notAD_train,notAD_predicted_as_AD_train] = evaluate_QDA(hippocampus_norm_AD_train,entorhinal_norm_AD_train,
                                                                           hippocampus_norm_notAD_train,entorhinal_norm_notAD_train,
                                                                           0.5)

print("Training set")
print("AD predicted as AD: " + str(AD_predicted_as_AD_train))
print("AD predicted as not AD: " + str(AD_predicted_as_notAD_train))
print("not AD predicted as not AD: " + str(notAD_predicted_as_notAD_train))
print("not AD predicted as AD: " + str(notAD_predicted_as_AD_train))
print("Error rate: " + str(error_rate_train))

step = 0.01
thresh_array = np.arange(0 + step,1 - step,step)
error_rate_array = np.zeros(len(thresh_array))
for ind,thresh in enumerate(thresh_array):
    [error_rate_train, AD_predicted_as_AD_train, AD_predicted_as_notAD_train,
     notAD_predicted_as_notAD_train,notAD_predicted_as_AD_train] = evaluate_QDA(hippocampus_norm_AD_train,entorhinal_norm_AD_train,
                                                                           hippocampus_norm_notAD_train,entorhinal_norm_notAD_train,
                                                                           thresh)
    error_rate_array[ind] = error_rate_train
    
plt.figure(figsize=(12,9))
plt.xticks([0,0.25,0.5,0.75,1],fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.plot(thresh_array,error_rate_array,color="black",lw=3)
#plt.xlim([xmin,xmax])
#plt.ylim([0,f_max])
plt.xlabel("Error rate", fontsize=font_size,labelpad=10)
plt.ylabel("Threshold probability", fontsize=font_size,labelpad=15)
#plt.savefig('plots/gaussian_conditional_' + str(ind) + '.pdf',bbox_inches="tight")
```
