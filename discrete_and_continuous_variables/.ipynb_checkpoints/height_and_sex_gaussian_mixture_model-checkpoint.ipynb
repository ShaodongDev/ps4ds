```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from scipy import stats
from scipy.stats import norm

data_men = pd.read_csv ("../data/ANSUR II MALE Public.csv", encoding='latin-1')
data_women = pd.read_csv ("../data/ANSUR II FEMALE Public.csv", encoding='latin-1')

font_size = 25
font_size_legend = 22
font_size_ticks = 25

stature_men = data_men['stature'] / 10.
stature_women = data_women['stature'] / 10.
stature = np.concatenate((stature_men,stature_women))

print(len(stature_men))
print(len(stature_women))
print(len(stature))
```

```python
n_iter = 100

# Initialize parameters
alpha = np.zeros(n_iter)
mu = np.zeros((n_iter,2))
sigma = np.zeros((n_iter,2))

alpha[0] = 0.5
mu[0,0] = np.mean(stature)
mu[0,1] = np.mean(stature)
sigma[0,0] = np.std(stature)
sigma[0,1] = np.std(stature)

gamma_0_aux = (1-alpha[0]) * norm.pdf(stature,mu[0,0],sigma[0,0])
gamma_1_aux = alpha[0] * norm.pdf(stature,mu[0,1],sigma[0,1])
gamma_sum = gamma_0_aux + gamma_1_aux
gamma_0 = gamma_0_aux / gamma_sum
gamma_1 = gamma_1_aux / gamma_sum 
print(gamma_1)
print(gamma_0)
# gamma = aux_gamma
    
for ind in range(n_iter):
    L_0 = (1-alpha[ind]) * norm.pdf(stature,mu[ind,0],sigma[ind,0])
    L_1 = alpha[ind] * norm.pdf(stature,mu[ind,1],sigma[ind,1])
    L = gamma_0_aux + gamma_1_aux
    gamma_0 = L_0 / L
    gamma_1 = L_1 / L_sum 
    
    loglikelihood = np.sum(np.log(L_sum))
    print("Iteration: " + str(ind) + " mu_0 = " + str(mu[ind,0]) + " sigma_0: " + str(sigma[ind,0])
          + " mu_1 = " + str(mu[ind,1]) + " sigma_1: " + str(sigma[ind,1])
          + " log-likelihood: " + loglikelihood)
    
    if ind < n_iter - 1:
        
        alpha[0] = 0.5
        mu[0,0] = np.mean(stature)
        mu[0,1] = np.mean(stature)
        sigma[0,0] = np.std(stature)
        sigma[0,1] = np.std(stature)
        
```

```python
def KDE_1D(data,x,bw=None):
    kernel = stats.gaussian_kde(data,bw_method=bw)
    kde = kernel(x)
    return kde

y_max = 0.07

bw_vals = [0.15,0.2,0.25]
hmin = stature_all.min()
hmax = stature_all.max()
h = np.linspace(hmin,hmax,500)

for index,bw in enumerate(bw_vals):
    kde_stature = KDE_1D(stature_all,h,bw=bw)
    kde_stature_men = KDE_1D(stature_men,h,bw=bw)
    kde_stature_women = KDE_1D(stature_women,h,bw=bw)

    plt.figure(figsize=(12,9))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.plot(h,kde_stature,color="black",lw=3)
    plt.xlim([hmin,hmax])
    plt.ylim([0,y_max])
    plt.ylabel(r"Marginal probability density function",fontsize=font_size,labelpad=20)
    plt.xlabel('Height (cm)',fontsize=font_size,labelpad=20)
    plt.savefig('plots/height_sex_marginal_' + str(index) + '.pdf',bbox_inches="tight")

    plt.figure(figsize=(12,9))
    plt.xticks(fontsize=font_size_ticks) 
    plt.yticks(fontsize=font_size_ticks)
    plt.plot(h,kde_stature_women,color="black",lw=3,ls="dashed",label="Women")
    plt.plot(h,kde_stature_men,color="black",lw=3,label="Men")
    plt.xlim([hmin,hmax])
    plt.ylim([0,y_max])
    plt.ylabel(r"Conditional probability density function",fontsize=font_size,labelpad=20)
    plt.xlabel('Height (cm)',fontsize=font_size,labelpad=20)
    plt.legend(fontsize=font_size)
    plt.savefig('plots/height_sex_marginal_conditional_' + str(index)+'.pdf',bbox_inches="tight")
```

```python
xmin = -0.6
xmax = 1.6
vals = [0,1]
ymax = 1

p_men = float(len(stature_men) ) / float(len(stature_all))
marginal_pmf = [1-p_men,p_men]
print(p_men)

plt.figure(figsize=(9,9))
plt.bar(vals,marginal_pmf, width = 0.4, color = "lightgray", edgecolor="black",linewidth = 2)
plt.xticks(np.arange(0, 1+1, 1),["Women","Men"])
plt.xticks(fontsize=font_size) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([0,ymax])
plt.xlim([xmin,xmax])
#plt.ylabel("Probability mass function",fontsize=font_size,labelpad = 30)
plt.xlabel("Sex",fontsize=font_size,labelpad = 15)
plt.savefig('plots/height_sex_marginal_pmf.pdf',bbox_inches="tight")
```

```python
mu_all = np.mean(stature_all)
sigma_all = np.std(stature_all)
mu_men = np.mean(stature_men)
sigma_men = np.std(stature_men)
mu_women = np.mean(stature_women)
sigma_women = np.std(stature_women)

print("Women -- Mean: " + str(np.round(mu_women,2)) + " Standard deviation: " + str(np.round(sigma_women,2)))
print("Men -- Mean: " + str(np.round(mu_men,2)) + " Standard deviation: " + str(np.round(sigma_men,2)))


n_bins = 30

plt.figure(figsize=(12,9))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.hist(stature_all,bins=n_bins,color='white', edgecolor='black',
          linewidth=1,density=True,label='Normalized histogram')
plt.plot(h,norm.pdf(h,mu_all,sigma_all),lw=3,color="black",label='Gaussian model')
plt.ylim([0,y_max])
plt.xlim([hmin,hmax])
plt.xlabel('$a$ (height in cms)',fontsize=font_size,labelpad=20)
plt.legend(fontsize=font_size)
plt.savefig('plots/height_sex_gaussian_all.pdf',bbox_inches="tight")

plt.figure(figsize=(9,9))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.hist(stature_men,bins=n_bins,color='white', edgecolor='black',
          linewidth=1,density=True,label='Normalized histogram')
plt.plot(h,norm.pdf(h,mu_men,sigma_men),lw=3,color="black",label='Gaussian model')
plt.ylim([0,y_max])
plt.xlim([hmin,hmax])
plt.xlabel('$a$ (height in cms)',fontsize=font_size,labelpad=20)
#plt.legend(fontsize=font_size)
plt.savefig('plots/height_sex_gaussian_men.pdf',bbox_inches="tight")

plt.figure(figsize=(9,9))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.hist(stature_women,bins=n_bins,color='white', edgecolor='black',
          linewidth=1,density=True,label='Normalized histogram')
plt.plot(h,norm.pdf(h,mu_women,sigma_women),lw=3,color="black",label='Gaussian model')
plt.ylim([0,y_max])
plt.xlim([hmin,hmax])
plt.xlabel('$a$ (height in cms)',fontsize=font_size,labelpad=20)
#plt.legend(fontsize=font_size)
plt.savefig('plots/height_sex_gaussian_women.pdf',bbox_inches="tight")

marginal_pdf = p_men*norm.pdf(h,mu_men,sigma_men)+(1-p_men)*norm.pdf(h,mu_women,sigma_women)

plt.figure(figsize=(12,9))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.hist(stature_all,bins=n_bins,color='white', edgecolor='black',
          linewidth=1,density=True,label='Normalized histogram')
plt.plot(h,marginal_pdf,lw=3,color="black",label='Mixture model')
plt.plot(h,(1-p_men)*norm.pdf(h,mu_women,sigma_women),lw=3,color="black",ls="dashed",label='Women')
plt.plot(h,p_men*norm.pdf(h,mu_men,sigma_men),lw=4,color="black",ls="dotted",label='Men')
plt.ylim([0,y_max])
plt.xlim([hmin,hmax])
plt.xlabel('$a$ (height in cms)',fontsize=font_size,labelpad=20)
plt.legend(fontsize=font_size)
plt.savefig('plots/height_sex_mixture_all.pdf',bbox_inches="tight")
```

```python
women_given_stature = (1-p_men)*norm.pdf(h,mu_women,sigma_women) / marginal_pdf
men_given_stature =  p_men*norm.pdf(h,mu_men,sigma_men) / marginal_pdf

plt.figure(figsize=(12,9))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
#sns.kdeplot(temperature_1,bw=bw,color="black",lw=3,gridsize=1000, legend = False)#,color='white', edgecolor='black',linewidth=2)
plt.plot(h,women_given_stature,color="black",lw=3,ls="dashed",label='P( Woman | Height = h )')
plt.plot(h,men_given_stature,color="black",lw=3,label='P( Man | Height = h )')
plt.xlim([hmin,hmax])
plt.ylim([-0.05,1.3])
plt.ylabel(r"Conditional probability mass function",fontsize=font_size,labelpad=20)
plt.xlabel('Height (h)',fontsize=font_size,labelpad=20)
plt.legend(fontsize=font_size)
plt.savefig('plots/height_sex_conditional_pmf.pdf',bbox_inches="tight")
```
